```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
df[:2]
```

<h3 style="color:#189AB4;font-size:40px;">Normal type of <strong>Pie Chart</strong></h3>


<p style="font-size:120%">
    We will first create a normal pie chart with the help of plotly.
</p>

```python
fig1 = px.pie(df, names='type', 
             height=300, width=600, 
             title='Type Overview',
             color_discrete_sequence=['#4c78a8', '#72b7b2'])
fig1.update_layout(plot_bgcolor='#fafafa', paper_bgcolor='#fafafa')
fig1.show()
```

<h3 style="color:#189AB4;font-size:40px;">Normal type of <strong>Donut Chart</strong></h3>


<p style="font-size:120%">
    We will first create a normal donut chart with the help of plotly.
</p>

```python
fig2 = px.pie(df, names='type', 
             height=300, width=600, 
             hole=0.7, 
             title='Type Overview',
                   color_discrete_sequence=['#4c78a8', '#72b7b2'])
fig2.update_traces(hovertemplate=None, textposition='outside', textinfo='percent+label', rotation=0)
fig2.update_layout(margin=dict(t=100, b=30, l=0, r=0), showlegend=False,
                        plot_bgcolor='#fafafa', paper_bgcolor='#fafafa',
                        title_font=dict(size=39, color='#555', family="Lato, sans-serif"),
                        font=dict(size=17, color='#8a8d93'),
                        hoverlabel=dict(bgcolor="#444", font_size=13, font_family="Lato, sans-serif"))
fig2.show()
```

<h3 style="color:#189AB4;font-size:40px;"><strong>Donut Chart with Annotation</strong></h3>


<p style="font-size:120%">
    We will now create a donut chart and add text in the middle of the chart. <br>
    First we will add the add_annotation parameter in the chart so we can have text in donut chart. <br>
</p>   <br>

<p style="font-size:120%"><strong>
    Explanation of parameters:</strong><br>
    x and y: Where we want to put our text in the chart it basically represent the point on x-axis and y-axis respectively.<br>
    showarrow: This parameters tell us if we want our arrow to show or hide.<br>
    text: This parameters let's us write the content which we want to put in our chart.
</p>

```python
fig3 = px.pie(df, names='type', 
             height=300, width=600, 
             hole=0.7, 
             color_discrete_sequence=['#4c78a8', '#72b7b2'])
fig3.update_traces(hovertemplate=None, textposition='outside',
                  textinfo='percent+label', rotation=0)
fig3.update_layout(margin=dict(t=40, b=35, l=0, r=0), showlegend=False,
                        plot_bgcolor='#fafafa', paper_bgcolor='#fafafa',
                        font=dict(size=17, color='#8a8d93'),
                        hoverlabel=dict(bgcolor="#444", font_size=13, font_family="Lato, sans-serif"))
fig3.add_annotation(dict(x=0.5, y=0.53,  align='center',
                        xref = "paper", yref = "paper",
                        showarrow = False, 
                        text="<span style='font-size: 35px; color=#555; font-family:Times New Roman'>Type ♥️</span>"))
fig3.show()
```

```python
df1 = pd.read_csv('../input/fashion-clothing-products-catalog/myntra_products_catalog.csv')

#
df1 = df1.replace({'Boys':'Men', 'Girls':'Women'})
df1 = df1[df1['Gender']!='Unisex Kids']

# Replacing null values to Missing
df1['PrimaryColor'] = df1['PrimaryColor'].replace(np.nan,"Missing")
```

<h3 style="color:#189AB4;font-size:40px;">Donut Chart with <strong>Annotation and Images</strong></h3>

<p style="font-size:120%">
    We will now create a donut chart and add Text + Image in the middle of the chart. <br>
    First we will add the add_annotation parameter in the chart so we can have text in donut chart. <br>
    Then we will add the add_layout_image parameter in the chart so we can have image in donut chart. 
</p>   <br>

<p style="font-size:120%"><strong>
    Explanation of parameters:</strong><br>
    x and y: Where we want to put our image in the chart it basically represent the point on x-axis and y-axis respectively.<br>
    source: This parameters let's us add the URL link of the image.<br>
</p>
    

```python
fig = px.pie(df1, names='Gender', 
             height=300, width=600, 
             hole=0.7, 
             color_discrete_sequence=['#4c78a8', '#72b7b2', '#6b92bc'])
fig.update_traces(hovertemplate=None, textposition='outside',
                  textinfo='percent+label', rotation=50)
fig.update_layout(margin=dict(t=50, b=30, l=0, r=0), showlegend=False,
                        plot_bgcolor='#fafafa', paper_bgcolor='#fafafa',
                        font=dict(size=17, color='#8a8d93'),
                        hoverlabel=dict(bgcolor="#444", font_size=13, font_family="Lato, sans-serif"))
fig.add_annotation(dict(x=0.5, y=0.4,  align='center',
                        xref = "paper", yref = "paper",
                        showarrow = False, font_size=22,
                        text="<span style='font-size: 35px; color=#555; font-family:Times New Roman'>Gender</span>"))
fig.add_layout_image(
    dict(
        source="https://i.imgur.com/3Cab96Z.jpg",
        xref="paper", yref="paper",
        x=0.48, y=0.48,
        sizex=0.3, sizey=0.25,
        xanchor="right", yanchor="bottom", sizing= "contain",
    )
)
fig.add_layout_image(
    dict(
        source="https://i.imgur.com/c6QKoDy.jpg",
        xref="paper", yref="paper",
        x=0.55, y=0.48,
        sizex=0.3, sizey=0.25,
        xanchor="right", yanchor="bottom", sizing= "contain",
    )
)
```

<h3 style="color:#189AB4;font-size:40px;"><strong>Another Way to Represent Data</strong></h3>

<p style="font-size:120%">
    The main idea for this chart is to use the insights and convey the message to user not in a chart form. Let's see how it's done.
</p>

<p style="font-size:120%"><strong>
    How to make this chart?</strong>   <br>
First thing we will need to make a scatter chart and then we will design the chart.<br>
I used the image and text annotation in a proper manner and created a infogrphic.
    </p>

```python
fig = px.scatter(x=[1,1.2,1.3], y=[0,0,0], 
                 color_discrete_sequence=['#fff'])
# styling
fig.update_xaxes(visible=False) 
fig.update_yaxes(visible=False) 
fig.update_traces(textposition='top center')
fig.update_layout(height=300, width=500, 
                  plot_bgcolor='#fff', paper_bgcolor='#fff',
                  margin=dict(b=0,r=50,l=50,t=110),
                  title={'text': "<span style='font-size:45px; color=#444; font-family:Times New Roman'>Gender Overview</span><br>For simplicity boys & girls are replaced to Men & Womens.",
                         'y':0.8, 'x':0.5, 
                         'xanchor': 'center', 'yanchor': 'top'},
                  font=dict(size=9, color='#666'),
                 )
fig.add_layout_image(
    dict(
        source="https://i.imgur.com/3Cab96Z.jpg",
        xref="paper", yref="paper",
        x=0.4, y=0.38,
        sizex=0.6, sizey=0.635,
        xanchor="right", yanchor="bottom", sizing= "contain",
    )
)
fig.add_layout_image(
    dict(
        source="https://i.imgur.com/c6QKoDy.jpg",
        xref="paper", yref="paper",
        x=0.7, y=0.38,
        sizex=0.6, sizey=0.65,
        xanchor="right", yanchor="bottom", sizing= "contain",
    )
)
fig.add_annotation(dict(x=0.37, y=0.25, ax=0, ay=0,
                    xref = "paper", yref = "paper", 
                    text= "<span style='font-size:21px'><b>44.7%</b></span><br><br><span style='font-size:15px'>Women</span>"
                  ))
fig.add_annotation(dict(x=0.68, y=0.25, ax=0, ay=0,
                    xref = "paper", yref = "paper", 
                    text= "<span style='font-size:21px'><b>45.7%</b></span><br><br><span style='font-size:15px'>Men</span>"
                  ))
```
