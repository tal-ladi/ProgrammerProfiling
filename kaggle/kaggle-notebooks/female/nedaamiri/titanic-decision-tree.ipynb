<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>Decision Trees are an important type of algorithm for predictive modeling machine learning.  </b></p> 
<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>The classical decision tree algorithms have been around for decades and modern variations like random forest are among the most powerful techniques available.  </b></p> 
<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>Classification and Regression Trees or `CART` for short is a term introduced by `Leo Breiman` to refer to Decision Tree algorithms that can be used for classification or regression predictive modeling problems. </b></p>
<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>Classically, this algorithm is referred to as decision trees, but on some platforms like R they are referred to by the more modern term CART.
 </b></p>

<p style = "font-size : 22px; color : #0909FF ; font-family : 'fantasy'; "><b> CART Model Representation</b></p>
<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>
The representation for the CART model is a binary tree.
This is your binary tree from algorithms and data structures, nothing too fancy. Each root node represents a single input variable (x) and a split point on that variable (assuming the variable is numeric).
The leaf nodes of the tree contain an output variable (y) which is used to make a prediction.
Given a new input, the tree is traversed by evaluating the specific input started at the root node of the tree.
 </b></p>
<p style = "font-size : 22px; color : #0909FF ; font-family : 'fantasy'; "><b> Some advantages of decision trees are:</b></p>
<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>

1. Simple to understand and to interpret. Trees can be visualised.

2. Requires little data preparation. 

3. Able to handle both numerical and categorical data.

4. Possible to validate a model using statistical tests. 

5. Performs well even if its assumptions are somewhat violated by the true model from which the data were generated.
 </b></p>
<p style = "font-size : 22px; color : #0909FF ; font-family : 'fantasy'; "><b> The disadvantages of decision trees include:</b></p>
<p style = "font-size : 20px; color : #191970 ; font-family : 'fantasy'; "><b>

1.Overfitting. Mechanisms such as pruning (not currently supported), setting the minimum number of samples required at a leaf node or setting the maximum depth of the tree are necessary to avoid this problem.

2.Decision trees can be unstable. Mitigant: Use decision trees within an ensemble.

3.Cannot guarantee to return the globally optimal decision tree. Mitigant: Training multiple trees in an ensemble learner

4.Decision tree learners create biased trees if some classes dominate. Recommendation: Balance the dataset prior to fitting
 </b></p>

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b> Features & Label
</b></p>

```python
X=df_train.drop('Survived',axis=1)
y=df_train['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b> Scalling The Features
</b></p>

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

scaler.fit(X_train)
```

```python
scaled_X_train= scaler.transform(X_train)
scaled_X_test= scaler.transform(X_test)
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b> Build the model
</b></p>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Predicting Test Data</b></p>

```python
y_pred = model.predict(X_test)
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Evaluate model</b></p>

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X_train.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Visualize the Tree</b></p>

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15,8),dpi=150)
plot_tree(model);
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Reporting Model Results</b></p>

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X_test.columns);
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Understanding Hyperparameters</b></p>

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Max Depth</b></p>

```python
help(DecisionTreeClassifier)
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

<p style = "font-size : 25px; color : #0909FF; font-family : 'fantasy'; "><b>Max Leaf Nodes</b></p>

```python
pruned_tree = DecisionTreeClassifier(max_leaf_nodes=3)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

<p style = "font-size : 25px; color : #0909FF ; font-family : 'fantasy'; "><b>Criterion</b></p>

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
