# Lesson 4 - Model Validation

```python
# measure the quality of the model (i.e. predictive accuracy: how close the model's predicted values are to actual values) using Mean Absolute Error (MAE)
# MAE = actual values - predicted values

from sklearn.metrics import mean_absolute_error

# wrong way:
# wrong way to measure predictive accuracy is to use the same data (i.e. training data) to build and evaluate the model
val_preds = model.predict(X)
mean_absolute_error(y, val_preds)
```

```python
# correct way:
# correct way to measure predictive accuracy is to use data (i.e. validation data) that was not used to build the model 

from sklearn.model_selection import train_test_split

# QUESTION: What is the purpose of "random_state = 0" 
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

# build a new model (i.e. define + fit a model) using the Training dataset
decisionTree_model = DecisionTreeRegressor()
decisionTree_model.fit(train_X, train_y)

# evaluate the model (i.e. predict target + measure MAE) by comparing the predicted values with the actual values
val_preds_dt = decisionTree_model.predict(val_X)
print(mean_absolute_error(val_y, val_preds_dt))
```

# Lesson 5 - Underfitting and Overfitting

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):

    # build model
    decisionTree_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    decisionTree_model.fit(train_X, train_y)
    
    # evaluate model
    val_preds_dt2 = decisionTree_model.predict(val_X)
    mae = mean_absolute_error(val_y, val_preds_dt2)
    return(mae)

# change the depths of tree by changing the values of max_leaf_nodes and compare the MAEs
for max_leaf_nodes in [5, 50, 100, 500, 1000, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
```

**comment:**
- Of the options listed above, 50 is the optimal number of leaves because MAE is the lowest.

- The sweet spot between underfitting and overfitting is when max_leaf_node=50. 
- The best model is when max_leaf_node=50 with MAE=27825.

# Lesson 6: Experiment with Different Model | Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# build a model
randomForest_model = RandomForestRegressor(random_state=1)
randomForest_model.fit(train_X, train_y)

# evaluate the model
val_preds_rf = randomForest_model.predict(val_X)
print(mean_absolute_error(val_y, val_preds_rf))
```

**comment:**
- The model performance has improved from MAE=27825 with Decision Tree Model to MAE=23009 with Random Forest Model

# Lesson X: Improve the Model | Random Forest

```python
# Update X by adding more features

features_updated = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd',
'MSSubClass',
'LotArea',
'OverallQual',
'OverallCond',
'YearBuilt',
'YearRemodAdd',
'1stFlrSF',
'2ndFlrSF',
'LowQualFinSF',
'GrLivArea',
'FullBath',
'HalfBath',
'BedroomAbvGr',
'KitchenAbvGr',
'TotRmsAbvGrd',
'Fireplaces',
'WoodDeckSF',
'OpenPorchSF',
'EnclosedPorch',
'3SsnPorch',
'ScreenPorch',
'PoolArea',
'MiscVal',
'MoSold',
'YrSold']

X = train_df[features_updated]

# split data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# build a model
rf_model2 = RandomForestRegressor(random_state=1)
rf_model2.fit(train_X, train_y)

# evaluate the model
val_preds_rf2 = rf_model2.predict(val_X)
mae_rf_model2 = mean_absolute_error(val_preds_rf2, val_y)

print("Validation MAE for Random Forest Model: {:,.0f}".format(mae_rf_model2))
```

# Lesson X: Make Predictions using the Test Data

```python
# to improve accuracy, build a new Random Forest model by fitting all the Training data
rf_modelF = RandomForestRegressor(random_state=1)
rf_modelF.fit(X, y)

# make predictions using the new data (i.e. Test data)
test_X = test_df[features_updated]
test_preds = rf_modelF.predict(test_X)
```

```python
# Run the code to save predictions in the format used for competition scoring

output = pd.DataFrame({'Id': test_df.Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```

# Key Takeaways
- Use the Training data to build a ML model (i.e. define + fit a model)
- Use the Validation data to evaluate the model (i.e. predict target + measure MAE)

- Characteristics of Decision Trees:
    >- A deep tree with lots of leaves will overfit because each prediction is coming from historical data from only the few houses at its leaf. 
    >- Ashallow tree with few leaves will perform poorly because it fails to capture as many distinctions in the raw data.
    
- Characteristics of Random Forests: 
    >- Random Forest uses many trees, and it makes a prediction by averaging the predictions of each component tree. 
    >- It generally has much better predictive accuracy than a single decision tree and it works well with default parameters.
