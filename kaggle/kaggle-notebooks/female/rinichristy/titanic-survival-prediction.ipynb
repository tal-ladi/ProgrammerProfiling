## **Data Cleaning**

### **Filling age column:**

Its seem better to fill the age column with the mean age per passenger class. 

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass',y='Age',data=titanic_train,palette='winter');
```

```python
titanic_train.groupby('Pclass').mean()
```

Define a function for filling age based on these mean Age per Pclass

```python
def fill_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 30
        else:
            return 25
    else:
        return Age
```

Applying fill_age function

```python
titanic_train['Age'] = titanic_train[['Age', 'Pclass']].apply (fill_age, axis = 1)
titanic_train.head()
```

Check Heatmap again

```python
plt.figure(figsize =(12,6))
sns.heatmap(titanic_train.isnull(), yticklabels = False, cbar = False, cmap ='cividis');
```

```python
titanic_train.isnull().sum()
```

Embarked column contains 2 missing values. Better drop those two rows.  Cabin column can be deleted completely. 

```python
titanic_train.drop('Cabin', axis =1, inplace = True)
```

```python
#Dropping missing embarked rows
titanic_train.dropna(inplace = True)
```

## **Building a Logistic Regression Classification model**

### **Train Test Split**

Splitting the titanic_train data into test and train. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(titanic_train.drop('Survived',axis=1), titanic_train['Survived'], test_size=0.30, random_state=101)
```

### **Training and predicting**

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter =10000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

### **Model Evaluation**

```python
from sklearn.metrics import accuracy_score
print('Accuracy of the Logistic Regression Model is ', accuracy_score(y_test,predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

## **Building a Support Vector Classifier(SVC) Model**

```python
from sklearn.svm import SVC
svcmodel = SVC()
svcmodel.fit(X_train,y_train)
predictions = svcmodel.predict(X_test)
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
print('Accuracy of the SVC Model is ', accuracy_score(y_test,predictions))
print('\n', '\n','Confusion Matrix of SVC Model:' '\n', confusion_matrix(y_test,predictions))
print('\n', '\n','Classification Report for SVC Model:' '\n',classification_report(y_test,predictions))
```

## **Building a Binary Regression Classification Model using Artificial Neural Network**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.constraints import max_norm
model = Sequential()
# input layer
model.add(Dense(50, input_dim =17,  activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(150, activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(150, activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.2))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='rmsprop')
model.fit(x=X_train, 
          y=y_train, 
          epochs=25,
          batch_size=256,
          validation_data=(X_test, y_test), 
          )
model.summary()
```

```python
losses = pd.DataFrame(model.history.history)
losses[['loss','val_loss']].plot();
```

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
print('Accuracy of the ANN Model is ', accuracy_score(y_test,predictions))
print('\n', '\n','Confusion Matrix of ANN Model:' '\n', confusion_matrix(y_test,predictions))
print('\n', '\n','Classification Report for ANN Model:' '\n',classification_report(y_test,predictions))
```

## **Building a XGBoost Classifier Model**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(titanic_train.drop('Survived',axis=1), titanic_train['Survived'], test_size=0.30, random_state=101)

from xgboost import XGBClassifier
model = XGBClassifier(learning_rate=0.05, n_estimators=200, objective='binary:logistic')
#model = XGBClassifier()
model.fit(X_train, y_train)

# Assess trained model performance on training dataset
predict_train = model.predict(X_train)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train, predict_train)
plt.figure()
sns.heatmap(cm, annot=True, cmap = 'plasma');
```

```python
# make predictions on test data
predict = model.predict(X_test)

# plot the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predict)
plt.figure()
sns.heatmap(cm, annot=True, cmap = 'plasma');
```

## **Prediction on Balanced Unbiased Data**

```python
plt.figure(figsize=(8,6))
sns.countplot(x = titanic_train.Survived);
```

```python
# Get all the Survived values separated.
Survivors = titanic_train[titanic_train.Survived == 1]
Victims = titanic_train[titanic_train.Survived== 0]
```

```python
len(Survivors)
```

```python
len(Victims)
```

```python
# Percentage of Survivors
(len(Survivors)/len(Victims))*100 # 61.9%
```

```python
# one way to fix it is to downsample the Victim rows
Victims_df = Victims.sample(n = len(Survivors), random_state = 40)
Survivors_df = Survivors
print(Survivors_df.shape, Victims_df.shape)
```

```python
# Create a dataframe with these Survivors and Victims counts
Survived_df = Survivors_df.append(Victims_df).reset_index(drop=True)
plt.figure(figsize=(8,6))
sns.countplot(x = Survived_df.Survived)
plt.title('Distribution of Survivors and Victims (after downsampling)')
plt.xlabel('Survived');
```

```python
X_train, X_test, y_train, y_test = train_test_split(Survived_df.drop('Survived',axis=1), Survived_df['Survived'], test_size=0.30, random_state=101)

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter =10000)
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
from sklearn.metrics import accuracy_score
print('Accuracy of LogisticRegression Model is ', accuracy_score(y_test,predictions))
print('\n', '\n','Confusion Matrix of LogisticRegression Model:' '\n', confusion_matrix(y_test,predictions))
print('\n', '\n','Classification Report for LogisticRegression Model:' '\n',classification_report(y_test,predictions))

# plot the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
plt.figure()
sns.heatmap(cm, annot=True, cmap = 'plasma');
```

## **Final Output of Prediction using Logistic Regression Classifier model on Balanced Unbiased Data**

```python
submission_preds = logmodel.predict(titanic_test)
test_ids = titanic_test['PassengerId']
df = pd.DataFrame({'PassengerId': test_ids.values, 'Survived': submission_preds})
df.to_csv('submission.csv', index = False)
```
