```python
train_data[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)

#keep as better class = more survival rate
```

```python
train_data['age_group'] = pd.cut(train_data['Age'], 4)

train_data[['age_group', 'Survived']].groupby(['age_group'], as_index=False).count().sort_values(by='Survived', ascending=True)
```

```python
for dataset in combine:
    dataset.loc[ dataset['Age'].isna(), 'Age'] = 0
    dataset.loc[ dataset['Age'] <= 20, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 20) & (dataset['Age'] <= 40), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 40) & (dataset['Age'] <= 60), 'Age']   = 2
    dataset.loc[ dataset['Age'] > 60, 'Age'] = 3
    dataset['Age'] = dataset['Age'].astype(int)
```

```python
train_data['FareBand'] = pd.qcut(train_data['Fare'], 4)
train_data[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).count().sort_values(by='FareBand', ascending=True)
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'].isna(), 'Fare'] = 0
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
train_data[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#siblings/spouse aboard

train_data[["SibSp", "Survived"]].groupby(['SibSp'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
# parents / children

train_data[["Parch", "Survived"]].groupby(['Parch'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_data = train_data.drop(['Name','Ticket', 'Cabin'], axis=1)
test_data = test_data.drop(['Name','Ticket', 'Cabin'], axis=1)
combine = [train_data, test_data]
```

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

```python
freq_port = train_data.Embarked.dropna().mode()[0]
freq_port

for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)
    
train_data[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

```python
x_test = test_data.drop(['PassengerId'],axis = 1)
x_test.head()
```

```python
train_data = train_data.drop(['age_group','FareBand'], axis=1)
```

```python
train_data.head()
```

```python
x_train = train_data.drop(['Survived', 'PassengerId'], axis = 1)
x_train
```

```python
y_train = train_data['Survived']
y_train
```

```python
x_train.shape, y_train.shape, x_test.shape
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators = 100)

random_forest.fit(x_train,y_train)
y_pred1 = random_forest.predict(x_test)

acc_random_forest = round(random_forest.score(x_train, y_train) * 100, 2)
acc_random_forest
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree.fit(x_train,y_train)

y_pred2 = decision_tree.predict(x_test)
acc_decision_tree = round(decision_tree.score(x_train,y_train) * 100, 2)
acc_decision_tree
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train,y_train)

y_pred3 = svc.predict(x_test)
acc_svc = round(svc.score(x_train,y_train)*100,2)
acc_svc
```
