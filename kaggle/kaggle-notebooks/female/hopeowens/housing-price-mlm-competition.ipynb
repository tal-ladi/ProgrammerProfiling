### <span style="color:#004c6d"> 3.2 Replacing missing values </span>  

These datasets had many missing values. It is important to account for missing values to minimize noise in the machine learning model. I needed to identify the type of missing data, the extent of the problem, and to verify the context of each value within the data description.

To do this, I first counted the missing values by factor in each dataset. 

```python
# count NA's test dataset
test_na_total = test.isnull().sum()
test_na_sort = test_na_total[test_na_total>0].sort_values(ascending=False)
sns.set_style("white")
f, ax = plt.subplots(figsize=(8, 7))
test_na_sort.plot.bar(color="b")
```

Figure 1: Frequency of missing values within the test dataset, (n = 1460). Blue bars show the frequency of missing values per variable.

```python
# count NA's train dataset
train_na_total = train.isnull().sum()
train_na_sort = train_na_total[train_na_total>0].sort_values(ascending=False)
sns.set_style("white")
f, ax = plt.subplots(figsize=(8, 6))
train_na_sort.plot.bar(color="b")
```

Figure 2: Frequency of missing values within the train dataset, (n = 1459). Blue bars show the frequency of missing values per variable.

```python
# replace missing values according to each variable's data description or imputation

#PoolQC- data description lists NA as "No Pool"
test["PoolQC"] = test["PoolQC"].fillna("None")
train["PoolQC"] = train["PoolQC"].fillna("None")

#MiscFeature- data description lists NA as "no misc feature"
test["MiscFeature"] = test["MiscFeature"].fillna("None")
train["MiscFeature"] = train["MiscFeature"].fillna("None")

#Alley- data description lists NA as "no alley access"
test["Alley"] = test["Alley"].fillna("None")
train["Alley"] = train["Alley"].fillna("None")

#Fence- data description lists NA as "no fence"
test["Fence"] = test["Fence"].fillna("None")
train["Fence"] = train["Fence"].fillna("None")

#FireplaceQu- data description lists NA as "no fireplace"
test["FireplaceQu"] = test["FireplaceQu"].fillna("None")
train["FireplaceQu"] = train["FireplaceQu"].fillna("None")

#LotFrontage- fill missing lot values by median lot frontage of the respective neighborhood.
test["LotFrontage"] = test.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
train["LotFrontage"] = train.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))

#GarageType, GarageFinish, GarageQual and GarageCond- Replacing missing data with None
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    test[col] = test[col].fillna('None')
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    train[col] = train[col].fillna('None')

#GarageYrBlt, GarageArea and GarageCars- no garage, replace missing data with 0
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    test[col] = test[col].fillna(0)
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    train[col] = train[col].fillna(0)

#BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF, BsmtFullBath and BsmtHalfBath- no basement, replace missing values with 0
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    test[col] = test[col].fillna(0)
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    train[col] = train[col].fillna(0)

#BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1 and BsmtFinType2- NaN lists as no basement
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    test[col] = test[col].fillna('None')
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    train[col] = train[col].fillna('None')

#MasVnrArea and MasVnrType- NA lists as no masonry veneer, area = 0, type = None
test["MasVnrType"] = test["MasVnrType"].fillna("None")
test["MasVnrArea"] = test["MasVnrArea"].fillna(0)
train["MasVnrType"] = train["MasVnrType"].fillna("None")
train["MasVnrArea"] = train["MasVnrArea"].fillna(0)

#MSZoning (The general zoning classification)- 
# Assumption: fill in with 'RL', most common zoning type
test['MSZoning'] = test['MSZoning'].fillna(test['MSZoning'].mode()[0])
train['MSZoning'] = train['MSZoning'].fillna(train['MSZoning'].mode()[0])

#Functional- data description lists NA as typical
test["Functional"] = test["Functional"].fillna("Typ")
train["Functional"] = train["Functional"].fillna("Typ")

#Electrical- Missing 1 value. Assumption: fill in with 'SBrkr', most common electrical type
test['Electrical'] = test['Electrical'].fillna(test['Electrical'].mode()[0])
train['Electrical'] = train['Electrical'].fillna(train['Electrical'].mode()[0])

#KitchenQual- Missing 1 value. Assumption: fill in with 'TA', most common KitchenQual type.
test['KitchenQual'] = test['KitchenQual'].fillna(test['KitchenQual'].mode()[0])
train['KitchenQual'] = train['KitchenQual'].fillna(train['KitchenQual'].mode()[0])

#Exterior1st and Exterior2nd- Missing 1 value. Assumption: fill in with most common exterior type
test['Exterior1st'] = test['Exterior1st'].fillna(test['Exterior1st'].mode()[0])
test['Exterior2nd'] = test['Exterior2nd'].fillna(test['Exterior2nd'].mode()[0])
train['Exterior1st'] = train['Exterior1st'].fillna(train['Exterior1st'].mode()[0])
train['Exterior2nd'] = train['Exterior2nd'].fillna(train['Exterior2nd'].mode()[0])

#SaleType- Missing 1 value. Assumption: fill in with "WD", the most frequent sale type
test['SaleType'] = test['SaleType'].fillna(test['SaleType'].mode()[0])
train['SaleType'] = train['SaleType'].fillna(train['SaleType'].mode()[0])

#MSSubClass- Na lists as no building class, replace missing values with None
test['MSSubClass'] = test['MSSubClass'].fillna("None")
train['MSSubClass'] = train['MSSubClass'].fillna("None")
```

### <span style="color:#004c6d">  6.3 Create optional log and square transformed features for the model  </span>

Manually calculate the log and square transformations of variables as potential factors.

```python
# Create logs function: for values l in list ls(dataframe), reassign log transformed values (res) in new columns 
# (add 1 to m/column), and end column with '_log'

def logs(res, ls):
    m = res.shape[1]
    for l in ls:
        res = res.assign(newcol=pd.Series(np.log(1.01+res[l])).values)   
        res.columns.values[m] = l + '_log'
        m += 1
    return res

# set list of features to log transform
log_features = ['LotFrontage','LotArea','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF',
                 'TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea',
                 'BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr',
                 'TotRmsAbvGrd','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF',
                 'EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','YearRemodAdd','TotalSF']

#all_features apply the logs function to all_features dataframe for log_features variables
all_features = logs(all_features, log_features)
```

```python
# Create squares function: for values l in list ls(dataframe), reassign squared values (res) in new columns 
# (add 1 to m/column), and end column with '_sq'

def squares(res, ls):
    m = res.shape[1]
    for l in ls:
        res = res.assign(newcol=pd.Series(res[l]*res[l]).values)   
        res.columns.values[m] = l + '_sq'
        m += 1
    return res 

# set list of features to transform
squared_features = ['YearRemodAdd', 'LotFrontage_log', 
              'TotalBsmtSF_log', '1stFlrSF_log', '2ndFlrSF_log', 'GrLivArea_log',
              'GarageCars_log', 'GarageArea_log']

#all_features apply the function to all_features dataframe for squared_features variables
all_features = squares(all_features, squared_features)
```

### <span style="color:#004c6d">  6.5 Model training </span>
Steps in the model training process:
* Cross Validation- a resampling method that uses different portions of the data to test and train a model on different iterations. This project uses a 12-fold cross-validation
* Models: using 7 hyper-parameter fitting models often used in kaggle data science comptetitions (as opposed to real life stats) 
    * **ridge**- regression model where the loss function is the linear least squares function and is given by the l2-norm. Also known as Ridge Regression or Tikhonov regularization.
    * **svr**- minimizes the l2-norm of the coefficients, not the the sums of squared. The model sets error constraints where the absolute error less than or equal to a specified margin, called the maximum error, ϵ (epsilon).
    * **gradient boosting**- combines weak regression trees into a single strong tree in an iterative fashion, "learning" from previous trees
    * **random forest**- combines weak regression trees into a single strong tree in an iterative fashion, selecting majority wins fitness at the end
    * **xgboost**-combines weak regression trees into a single strong tree based on minimizing the loss function
    * **lightgbm regressors**- combines weak regression trees into a single strong tree building "leaf wise" not "base wise"
* Stacking and Blending: meta-learning algorithms that combine predictions from two or more base machine learning algorithms to create better performance fit to a specific data source.

```python
#import sklearn.model_selection
#sklearn.model_selection.KFold
from sklearn.model_selection import KFold
KFold
# Setup cross validation folds- 12 fold cross validation
kf = KFold(n_splits=12, random_state=42, shuffle=True)
```

```python
# Define error metrics to fit models
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

def cv_rmse(model, X=X):
    rmse = np.sqrt(-cross_val_score(model, X, train_labels, scoring="neg_mean_squared_error", cv=kf))
    return (rmse)
```

```python
# Run machine learning models
# Light Gradient Boosting Regressor
lightgbm = LGBMRegressor(objective='regression', 
                       num_leaves=6,
                       learning_rate=0.01, 
                       n_estimators=7000,
                       max_bin=200, 
                       bagging_fraction=0.8,
                       bagging_freq=4, 
                       bagging_seed=8,
                       feature_fraction=0.2,
                       feature_fraction_seed=8,
                       min_sum_hessian_in_leaf = 11,
                       verbose=-1,
                       random_state=42)

# XGBoost Regressor
xgboost = XGBRegressor(learning_rate=0.01,
                       n_estimators=6000,
                       max_depth=4,
                       min_child_weight=0,
                       gamma=0.6,
                       subsample=0.7,
                       colsample_bytree=0.7,
                       objective='reg:linear',
                       nthread=-1,
                       scale_pos_weight=1,
                       seed=27,
                       reg_alpha=0.00006,
                       random_state=42)

# Ridge Regressor
ridge_alphas = [1e-15, 1e-10, 1e-8, 9e-4, 7e-4, 5e-4, 3e-4, 1e-4, 1e-3, 5e-2, 1e-2, 0.1, 0.3, 1, 3, 5, 10, 15, 18, 20, 30, 50, 75, 100]
ridge = make_pipeline(RobustScaler(), RidgeCV(alphas=ridge_alphas, cv=kf))

# Support Vector Regressor
svr = make_pipeline(RobustScaler(), SVR(C= 20, epsilon= 0.008, gamma=0.0003))

# Gradient Boosting Regressor
gbr = GradientBoostingRegressor(n_estimators=6000,
                                learning_rate=0.01,
                                max_depth=4,
                                max_features='sqrt',
                                min_samples_leaf=15,
                                min_samples_split=10,
                                loss='huber',
                                random_state=42)  

# Random Forest Regressor
rf = RandomForestRegressor(n_estimators=1200,
                          max_depth=15,
                          min_samples_split=5,
                          min_samples_leaf=5,
                          max_features=None,
                          oob_score=True,
                          random_state=42)

# Stack models and optimize using xgboost
stack_gen = StackingCVRegressor(regressors=(xgboost, lightgbm, svr, ridge, gbr, rf),
                                meta_regressor=xgboost,
                                use_features_in_secondary=True)
```

### <span style="color:#004c6d">  6.6 Testing model accuracy  </span> 
#### <span style="color:#004c6d">  RMSE Values  </span>  
Our training dataset now has both the model predictions and known sale price values listed as separate columns. Using this, we can statistically test the accuracy of model predictions. Our next goal is to find the most accurate model, and use it to predict sales price in the test (unknown values) dataset. 

Root Mean Square Error (RMSE) is the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data points are; RMSE is a measure of how spread out these residuals are. In other words, it tells you how concentrated the data is around the line of best fit. Root mean square error is commonly used in climatology, forecasting, and regression analysis to verify experimental results.

${\texttt{rmse} = \bigg\{\sqrt{(\frac{1}{n})\sum_{i=1}^{n}(y_{i} - x_{i})^{2}} \bigg\}\tag{1}}$

Models with lower rmse-score values, ranging from 0 to 1, have higher accuracy in predicting sales price of houses.
