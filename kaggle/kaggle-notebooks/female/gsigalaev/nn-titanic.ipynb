# Neural Network - Titanic

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, OneHotEncoder

# Read in the training and test sets
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
test_df = pd.read_csv('/kaggle/input/titanic/test.csv')
train_df.head()
```

```python
relevant_features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

X = train_df [relevant_features]
y = train_df["Survived"]

X.head()
```

```python

categorial = [col for col in X.select_dtypes(include="object")]
numerical = [col for col in X.select_dtypes(exclude ="object")]

numerical_pipeline = Pipeline( steps = [
                            ("imputer", SimpleImputer(strategy="most_frequent"))
])

categorial_pipeline = Pipeline(steps =[
                            ("Imputer", SimpleImputer(strategy = "most_frequent")),
                            ("encode", OneHotEncoder(handle_unknown ="ignore"))
])

pp = ColumnTransformer(transformers =[
                            ("num",numerical_pipeline, numerical),
                            ("cat", categorial_pipeline, categorial)
])


X = pp.fit_transform(X)
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

X_train.shape
```

```python
# using Subclassing API
import tensorflow as tf
from tensorflow import keras

class Basic_NN_Model(keras.models.Model):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        
        # define the NN architecture#
        self.hidden1 = keras.layers.Dense(50, activation='relu', input_shape=(10,))
        self.hidden2 = keras.layers.Dense(50, activation='relu')
        self.main_output = keras.layers.Dense(1, activation='sigmoid')        
        
    def call(self, inputs):
        hidden1 = self.hidden1(inputs)
        hidden2 = self.hidden2(hidden1)
        main_output = self.main_output(hidden2)
        
        return main_output
    
model = Basic_NN_Model()

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=40, validation_data=(X_test, y_test))

test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print("Test Accuracy: {:.2f}%".format(test_accuracy * 100))
```
