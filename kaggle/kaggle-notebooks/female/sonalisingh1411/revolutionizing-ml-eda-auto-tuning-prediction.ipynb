<p style= "font-family:Georgia;color:#000000;font-size:110%;text-align:center;">Introduction About <b>Sherpa</b><br>
    <br>
Sherpa can automatically run parallel evaluations on a cluster using a job scheduler such as SGE. Simply provide a Python script that takes a set of hyperparameters as arguments and performs a single trial evaluation. A database collects the partial results in real-time, and the hyperparameter optimization algorithm decides what to do next.<br>
<br>
SHERPA is a Python library for hyperparameter tuning of machine learning models. It provides:<br>
<br>
Hyperparameter optimization for machine learning researchers.<br>
1.It can be used with any Python machine learning library such as Keras, Tensorflow, PyTorch, or Scikit-Learn.<br>
2.A choice of hyperparameter optimization algorithms such as Bayesian optimization via GPyOpt, Asynchronous Successive Halving (aka Hyperband) , and Population Based Training .<br>
4.Parallel computation that can be fitted to the user's needs.<br>
5.A live dashboard for the exploratory analysis of results.<br>

```python
from sklearn.model_selection import cross_val_score
import time
import sherpa
import sherpa.algorithms.bayesian_optimization as bayesian_optimization
```

```python
parameters = [sherpa.Discrete('n_estimators', [2, 50]),
              sherpa.Choice('criterion', ['gini', 'entropy']),
              sherpa.Continuous('max_features', [0.1, 0.9])]

algorithm = bayesian_optimization.GPyOpt(max_concurrent=1,model_type='GP_MCMC',acquisition_type='EI_MCMC',max_num_trials=20)
```

```python
%%time
study = sherpa.Study(parameters=parameters,
                     algorithm=algorithm,
                     lower_is_better=False)

for trial in study:
    print("Trial ", trial.id, " with parameters ", trial.parameters)
    clf = RandomForestClassifier(criterion=trial.parameters['criterion'],
                                 max_features=trial.parameters['max_features'],
                                 n_estimators=trial.parameters['n_estimators'],
                                 random_state=0)
    scores = cross_val_score(clf, x, Y, cv=5)
    print("Score: ", scores.mean())
    study.add_observation(trial, iteration=1, objective=scores.mean())
    study.finalize(trial)
print(study.get_best_result())
```

<p style= "font-family:Georgia;color:#000000;font-size:110%;text-align:left;"><br>
🎯<b>Confusion Matrix</b> is a tool to determine the performance of the classifier. It contains information about actual and predicted classifications. The below table shows the confusion matrix of the two-class:-<br>
<b>Class: 1</b> - The person has diabetes.<br>
<b>Class: 0</b> - The person doesn't have diabetes.<br>
<br>
    <br>
🎯 <b>True Positive (TP)</b> is the number of correct predictions that an example is positive which means the positive class is correctly identified as positive. Example: Given class say that <b>the person has diabetes</b> and <b>the classifier has been correctly predicted that the person has diabetes.</b><br>
<br>
🎯 <b>False Negative (FN)</b> is the number of incorrect predictions that an example is negative which means the positive class is incorrectly identified as negative. Example: Given class say that <b>the person has diabetes</b>, however,  <b>the classifier has been incorrectly predicted that the person doesn't have diabetes.</b><br>

<br>
🎯 <b>False positive (FP)</b> is the number of incorrect predictions that an example is positive which means the negative class is incorrectly identified as positive. Example: Given class say that <b>the person doesn't have diabetes</b>, however, <b>the classifier has been incorrectly predicted that the person has diabetes.</b><br>
<br>
🎯 <b>True Negative (TN)</b> is the number of correct predictions that an example is negative which means the negative class is correctly identified as negative. Example: Given class that, <b>the person doesn't have diabetes</b> and <b>the classifier has been correctly predicted that the person doesn't have diabetes</b><br>
</p>
