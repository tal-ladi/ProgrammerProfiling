# Import Libaries

```python
import numpy as np
import pandas as pd
```

# Loading Training dataset

```python
train_data=pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
train_data
```

# Slicing Training dataset

```python
x_data=train_data.iloc[:,:-1]
x_data
```

# Converting string data inti numeric data

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
for col in train_data.columns:
    if train_data[col].dtype == object:
        train_data[col] = train_data[col].astype(str)
        d1=train_data[col].unique()
        d2=d1[-2]
        
        train_data[col]=train_data[col].fillna(d2)
        train_data[col]=lb.fit_transform(train_data[col])
    elif train_data[col].dtype == float or train_data[col].dtype == int:
        m1=train_data[col].mean()
        m1=round(m1)
        train_data[col]=train_data[col].fillna(m1)
        
print(train_data)
```

```python
del train_data["Transported"]
```

```python
train_data
```

```python
test_data=pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
test_data
```

```python
for col in test_data.columns:
    if test_data[col].dtype == object:
        test_data[col] = test_data[col].astype(str)
        d1=test_data[col].unique()
        d2=d1[-2]
        
        test_data[col]=test_data[col].fillna(d2)
        test_data[col]=lb.fit_transform(test_data[col])
    elif test_data[col].dtype == float or test_data[col].dtype == int:
        m1=test_data[col].mean()
        m1=round(m1)
        test_data[col]=test_data[col].fillna(m1)
        
print(test_data)
```

# Loading Training dataset

```python
train_data1=pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
train_data1
```

# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
lb1=LabelEncoder()
train_data1["Transported"]=lb1.fit_transform(train_data1["Transported"])
```

```python
train_data1["Transported"]
```

# Slicing  Training Dataset

```python
x=train_data.iloc[:,:].values
y=train_data1.iloc[:,-1].values
y
```

```python
test_std=test_data.iloc[:,:].values
```

# Using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x=std.fit_transform(x)
test_std=std.fit_transform(test_std)
```

```python
test_std.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)
```

# Using Logistic regression

```python
from sklearn.linear_model import LogisticRegression 
log=LogisticRegression()
log.fit(x_train,y_train)

y_pred=log.predict(x_test)
y_pred

from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)*100
ac
```

# Using Support Vector Machine(SVM)

```python
from sklearn.svm import SVC
svc1=SVC(kernel='linear',random_state=11)
svc1.fit(x_train,y_train)

y_pred1=svc1.predict(x_test)
y_pred1

from sklearn.metrics import accuracy_score
ac1=accuracy_score(y_test,y_pred1)*100
print(ac1)
```

```python
from sklearn.svm import SVC
svc2=SVC(kernel='linear',random_state=11)
svc2.fit(x,y)

final_pred=svc2.predict(test_std)
final_pred
```

```python
final=[]
for x in final_pred:
    if x==0:
        final.append(False)
    else:
        final.append(True)
```

```python
final1=pd.read_csv(r"/kaggle/input/spaceship-titanic/test.csv")
final1.insert(1,"Transported",final)
final2=final1.iloc[:,[0,1]]
final2
```
