# Visualization

## ScatterPlot

```python
# We will make a scatterplot of two of the four Iris features 


iris.plot(kind="scatter", x="SepalLengthCm", y= "SepalWidthCm")
```

## JointPlot

```python
## we can do it also with the seaborn library

## A seaborn jointplot shows bivariate scatterplots and univariate histograms in the 
# same figure

sns.jointplot(data= iris, x="SepalLengthCm", y="SepalWidthCm", size=5)
```

## FacetGrid

```python

sns.FacetGrid(iris, hue="Species", palette="husl",size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

## BoxPlot

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

## Violin Plot

```python
sns.violinplot(data=iris, x="Species", y="PetalWidthCm",
                palette="rocket")
```

## PairPlot

```python
sns.pairplot(iris.drop("Id", axis=1), hue ="Species",palette= "husl", size=3) 
 
```
