```python
import pandas as pd
import numpy as np
from sklearn import feature_extraction, linear_model, model_selection, preprocessing
```

```python
train = pd.read_csv("../input/nlp-getting-started/train.csv")
test = pd.read_csv("../input/nlp-getting-started/test.csv")
```

Data set contains keyword, location, text of a tweet and target, which indicates, whether a tweet is about disaster or not. The point of this exercise is to train a model which predicts the target.

```python
train.head()
```

### Transforming words into vectors

Vectors contain information on how many times each word from a tweet occurs in the whole set of tweets.

```python
tfidf_vetorizer = feature_extraction.text.TfidfVectorizer()
train_vectors = tfidf_vetorizer.fit_transform(train["text"])
test_vectors = tfidf_vetorizer.transform(test["text"])
```

```python
print(train_vectors)
```

### Creating the model

Since there is linear connection between particular words and the indication, if the tweet is about disaster or not, the model used is Ridge Classifier.

Diffrent possible approaches: LSA, LSTM / RNNs

```python
model = linear_model.RidgeClassifier()

#prediction
model.fit(train_vectors, train["target"])
```

```python
train["target"]
```

### Evaluation

Cross-validation is used in order to validate the quality of the model. 

```python
scores = model_selection.cross_val_score(model, train_vectors, train["target"], scoring="f1")
```

```python
np.average(scores)
```

### Submission

```python
test["target"] = model.predict(test_vectors)
```
