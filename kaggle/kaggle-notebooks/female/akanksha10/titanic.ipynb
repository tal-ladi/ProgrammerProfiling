# Reading Csv File

```python
df1=pd.read_csv("/kaggle/input/titanic/gender_submission.csv")
df2=pd.read_csv("/kaggle/input/titanic/train.csv")
df3=pd.read_csv("/kaggle/input/titanic/test.csv")
```

# Cleaning The Data

```python
df2=df2.drop(columns=["PassengerId","Name","Ticket","Fare","Cabin","Embarked"])
```

```python
df3=df3.drop(columns=["PassengerId","Name","Ticket","Fare","Cabin","Embarked"])
```

```python
df2.describe()
```

```python
df3.describe()
```

```python
df2['Age'].mean()
df2['Age'].fillna(df2['Age'].mean(),inplace=True)
```

```python
df3['Age'].mean()
df3['Age'].fillna(df3['Age'].mean(),inplace=True)
```

```python
dummies=pd.get_dummies(df2['Sex'])
dummies.head()
```

```python
df2
```

```python
df2=pd.concat([df2,dummies],axis='columns')
df2.head()
```

```python
df2.drop(['Sex','female'],axis=1,inplace=True)
df2.head()
```

```python
cor=df2.corr()
cor
```

```python
x_df2=df2.iloc[:,2:]
y_df2=df2.Survived
```

```python
x_df2
```

```python
y_df2
```

```python
dummies=pd.get_dummies(df3['Sex'])
dummies.head()
```

```python
df3=pd.concat([df3,dummies],axis='columns')
df3.head()
```

```python
df3.drop(['Sex','female'],axis=1,inplace=True)
df3.head()
```

```python
df3.corr()
```

```python
x_df3=df3.iloc[:,1:]
```

```python
x_df2=x_df2.astype(int)
```

# Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_df2,y_df2)
```

```python
y_predict=model.predict(x_df3)
y_predict
```

# Got 100% Accuracy

```python
model.score(x_df3,y_predict)*100
```
