# Load The Data

```python
#Read Training Data
train_data=pd.read_csv('../input/titanic/train.csv')
#view the first 5 row
train_data.head()
```

```python
#view the Last 5 row from Training data
train_data.tail()
```

```python
#Read the Testting Data
test_data = pd.read_csv('../input/titanic/test.csv')
#view the first 5 row from the test Data
test_data.head()
```

```python
#View the last 5 row from the test data
test_data.tail()
```

# the rate of women who survived 

```python
#print the rate of women who survived 
women = train_data.loc[train_data.Sex=='female']["Survived"]
rate_women = sum(women)/len(women)
print('The Rate of Women who Survived is : ',rate_women , '%')
```

# the rate of Men who survived

```python
#print the rate of Men who survived
Men = train_data.loc[train_data.Sex=='male']["Survived"]
rate_men = sum(Men)/len(Men)
print('The Rate of Men who Survived is :' , rate_men , '%')
```

# Fit The Model 

```python
from sklearn.ensemble import RandomForestClassifier
# y is training data target
y=train_data["Survived"]
features = ["Pclass", "Sex", "SibSp", "Parch"]
# X is training data features
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
#N_estimators is the numer of trees
#Max_depth is the deepth of Requared
R_F_C = RandomForestClassifier(n_estimators=100 , max_depth=5 , random_state=1)

#fitting the model
R_F_C.fit(X,y)
```

# Predictions

```python
predictions = R_F_C.predict(X_test)
```

# print PassengerId and prediction Survived or not

```python
output = pd.DataFrame({'PassengerId':test_data.PassengerId , 'Survived':predictions})
```
