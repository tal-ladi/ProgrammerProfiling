```python
# Splitting the dataset into train and test sets

X = df.iloc[:, :-1].to_numpy()
y = df['Class'].to_numpy()
```

```python
# Generate and plot a synthetic imbalanced classification dataset

counter = Counter(y)
print(counter)

# Scatter plot of 'Class' by class label
for label, _ in counter.items():
    row_ix = np.where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
# Transform the dataset

sm = SMOTE(random_state=0)
X, y = sm.fit_resample(X, y)
```

```python
counter = Counter(y)
print(counter)

# Scatter plot of 'Class' by class label
for label, _ in counter.items():
    row_ix = np.where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
# Split dataset into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

## Model

### 1. Decision Tree Classifier

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
print("Score of train data:", dt.score(X_train, y_train))
print("Score of test data:", dt.score(X_test, y_test))
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
print("Accuracy Score:", accuracy_score(y_pred, y_test))
```

```python
y_pred_prob = dt.predict_proba(X_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision Recall Curve')
```

```python
# Confusion Matrix

cm = confusion_matrix(y_test, y_pred)

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax=ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
```

### 2. Neural Netowrk

```python
import keras
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Initialising the ANN
classifier = Sequential()
classifier.add(Dense(units=30, kernel_initializer='he_uniform', activation='relu', input_dim = 30))
classifier.add(Dense(units=50, kernel_initializer='he_uniform', activation='relu'))
classifier.add(Dense(units=35, kernel_initializer='he_uniform', activation='relu'))
classifier.add(Dense(units=2, activation='softmax'))

# Compiling the ANN
classifier.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
# Fitting the ANN to the Training set

model_history = classifier.fit(X_train, y_train, validation_split=0.33, batch_size=250, epochs=30)
```

```python
# List all data in history
print(model_history.history.keys())

# Summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Summarize history for loss

plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Calculating score

score = classifier.evaluate(X_test, y_test)
score
```

```python
# Predicting on test data

y_pred = classifier.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
y_pred
```

```python
# Accuracy

accuracy_score(y_pred, y_test)
```

```python
# Confusion Matrix

cm = confusion_matrix(y_test, y_pred)

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax=ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
```
