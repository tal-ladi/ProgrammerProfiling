```python
import gensim
from gensim.models import Word2Vec 
from gensim.models import KeyedVectors
import pandas as pd
from nltk.tokenize import RegexpTokenizer
```

## Preprocessing

```python
# read in data with kaggle forum posts
forum_posts = pd.read_csv("../input/meta-kaggle/ForumMessages.csv")

# grab forum forum posts
sentences = forum_posts.Message.astype('str').tolist()

# tokenize
tokenizer = RegexpTokenizer(r'\w+')
sentences_tokenized = [w.lower() for w in sentences]
sentences_tokenized = [tokenizer.tokenize(i) for i in sentences_tokenized]
```

## Fine tuning

```python
# update existing embedding w/ kaggle data
model_2 = Word2Vec(size=300, min_count=1)
model_2.build_vocab(sentences_tokenized)
total_examples = model_2.corpus_count
model_2.intersect_word2vec_format("../input/word2vec-google/GoogleNews-vectors-negative300.bin", binary=True, lockf=1.0)
model_2.train(sentences_tokenized, total_examples=total_examples, epochs=5)
```

## Save tuned model

```python
# gensim flavored word2vec model (smaller)
model_2.save("kaggle_word2vec_gensim.model")

# generic word2vec model
model_2.wv.save_word2vec_format("kaggle_word2vec.model")
```
