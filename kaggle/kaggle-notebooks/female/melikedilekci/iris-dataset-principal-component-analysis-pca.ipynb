```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
data=iris.data
data
```

```python
feature_names=iris.feature_names
feature_names
```

```python
y=iris.target
y
```

```python
df=pd.DataFrame(data,columns=feature_names)
```

```python
df["sinif"]=y
```

```python
x=data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2,whiten=True)
```

```python
pca.fit(x)
```

```python
x_pca=pca.transform(x)
```

```python
print("varience ratio:",pca.explained_variance_ratio_)
```

```python
print("sum:",sum(pca.explained_variance_ratio_))
```

```python
# 2D visualization
```

```python
df["p1"]=x_pca[:,1]
df["p2"]=x_pca[:,1]
```

```python
color=["red","green","blue"]
```

```python
for each in range(3):
    plt.scatter(df.p1[df.sinif==each],df.p2[df.sinif==each],color=color[each],label=iris.target_names[each])
plt.legend()
plt.xlabel("p1")
plt.ylabel("p2")
plt.show()
    
    
```

```python
iris=load_iris()
```

```python
x=iris.data
```

```python
y=iris.target
```

```python
# normalization

x=(x-np.min(x)/(np.max(x)-np.min(x)))
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=3)
```

```python
#k=10

from sklearn.model_selection import cross_val_score
accuracies=cross_val_score(estimator=knn,X=x_train,y=y_train,cv=10)
accuracies
```

```python
print("average accuracies: ",np.mean(accuracies))
print("std accuracies: ",np.std(accuracies))
```

```python
knn.fit(x_train,y_train)
print("test accuracy:",knn.score(x_test,y_test))
```

```python
# grid search cross validation

from sklearn.model_selection import GridSearchCV

grid={"n_neighbors":np.arange(1,50)}
```

```python
knn=KNeighborsClassifier()
```

```python
knn_cv=GridSearchCV(knn,grid,cv=10)
```

```python
knn_cv
```

```python
knn_cv.fit(x,y)
```

```python
print("tuned hyperpatameter K:",knn_cv.best_params_)
print("best parameter best accursacy: ", knn_cv.best_score_)
```

```python
## grid search cv with logistic regression

x=x[:100,:]
y=y[:100]
```

```python

from sklearn.linear_model import LogisticRegression
```

```python
grid= {"C" :np.logspace(-3,3,7),"penalty":["l1","l2"]}
```

```python
logred=LogisticRegression()
```

```python
logred_cv=GridSearchCV(logred,grid,cv=10)

```

```python
logred_cv.fit(x,y)
```

```python
print("tuned hyperparameters: (best parameters)",logred_cv.best_params_)
```

```python
print("accuracy: (best parameters)",logred_cv.best_score_)
```
