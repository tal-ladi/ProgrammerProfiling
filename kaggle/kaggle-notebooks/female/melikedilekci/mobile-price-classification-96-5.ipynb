<a id="1"></a>
<h1 style="background-color:orange;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">Import Libraries</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<a id="2"></a>
<h1 style="background-color:orange;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">Load the Data</h1>

```python
train=pd.read_csv("/kaggle/input/mobile-price-classification/train.csv")
train.head()
```

<a id="3"></a>
<h1 style="background-color:ORANGE;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">Logistic Regression</h1>

```python
train = train[np.isfinite(train).all(1)]
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
```

```python
x_data=train.drop(["price_range"],axis=1)
y=train.price_range.values
```

```python
x=(x_data-np.min(x_data)/(np.max(x_data)-np.min(x_data))).values   # normalization
x
```

```python
test=pd.read_csv("/kaggle/input/mobile-price-classification/test.csv")
```

```python
test.shape
```

```python
train.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=42)
```

```python
print("x_train",len(x_train))
print("x_test",len(x_test))
print("y_train",len(y_train))
print("y_test",len(y_test))
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
acc_log_train = round(lr.score(x_train, y_train)*100,2) 
acc_log_test = round(lr.score(x_test,y_test)*100,2)
print("Training Accuracy: % {}".format(acc_log_train))
print("Testing Accuracy: % {}".format(acc_log_test))
```

<a id="4"></a>
<h1 style="background-color:ORANGE;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">KNN Classifier </h1>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_train,y_train)
```

```python
pred=knn.predict(x_test)
```

```python
print("{} nın score: {} ".format(10,knn.score(x_test,y_test)))
```

<a id="5"></a>
<h1 style="background-color:ORANGE;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">SVM</h1>

```python
from sklearn.svm import SVC
```

```python
svm=SVC(random_state=1)
svm.fit(x_train,y_train)
```

```python
print("print svm accuracy:", svm.score(x_test,y_test))
```

<a id="6"></a>
<h1 style="background-color:ORANGE;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">Naive Bayes</h1>

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB

nb=GaussianNB()
nb1=BernoulliNB()
nb3=CategoricalNB()
```

```python
nb.fit(x_train,y_train)
nb1.fit(x_train,y_train)
nb3.fit(x_train,y_train)
```

```python
print("GaussianNB accuracy:", nb.score(x_test,y_test))
print("BernoulliNB accuracy:", nb1.score(x_test,y_test))
print("CategoricalNB accuracy:", nb3.score(x_test,y_test))
```

<a id="7"></a>
<h1 style="background-color:ORANGE;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">Decision Tree</h1>

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()

dt.fit(x_train,y_train)
```

```python
print("Decision Tree Classifier Accuracy: ", dt.score(x_test,y_test))
```

<a id="8"></a>
<h1 style="background-color:ORANGE;font-family:newtimeroman;font-size:250%;text-align:center;border-radius: 10px 10px;">Random Forest</h1>

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100, random_state=1)
```

```python
rf.fit(x_train,y_train)
```

```python
print("Random Forest Accuracy: ", rf.score(x_test,y_test))
```
