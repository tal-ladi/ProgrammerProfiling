```python
import pandas as pd
import numpy as np

train = pd.read_csv("../input/titanic/train.csv")
test = pd.read_csv("../input/titanic/test.csv")
train = train.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1)
test = test.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1)
train = pd.concat([train, test], axis=0, ignore_index=True)
train
```

```python
train["is_rich"] = 0
train["is_young"] = 0
train["is_old"] = 0
train.loc[(train["Pclass"] == 1) | (train["Fare"] > 112), "is_rich"] = 1
train.loc[train["Age"] < 18, "is_young"] = 1
train.loc[train["Age"] > 60, "is_old"] = 1

enc_col = ["Embarked", "Pclass", "Sex", "is_rich", "is_young", "is_old"]
for col in enc_col:
    train[col + "_enc"] = train[col].map(train.groupby(col).size() / len(train))
train
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = ["Pclass", "Sex", "Embarked", "is_rich", "is_young", "is_old"]
ignore_col = ["Pclass", "is_rich", "is_young", "is_old"]
train.fillna("NaN", inplace=True)
for col in categorical_cols:
    if col in ignore_col:
        continue
    train[col] = LabelEncoder().fit_transform(train[col])
train
```

```python
columns = train.columns[1:]
columns
```

```python
from catboost import Pool, CatBoostClassifier, CatBoostRegressor
from sklearn.model_selection import train_test_split

test = train[891:].reset_index(drop=True)
train = train[:891].reset_index(drop=True)
train = train.astype({"Survived": "int8"})
test = test.drop("Survived", axis=1)

NUM_STACK_ROUND = 7
# pred = np.zeros((NUM_STACK_ROUND, len(train), 1))
# test_pred = np.zeros((NUM_STACK_ROUND, len(test), 1))
params = dict(
    grow_policy = "Lossguide",
    l2_leaf_reg = 2,
)
for i in range(NUM_STACK_ROUND):
    if i == 0:
        X_train, X_val, y_train, y_val = train_test_split(train[columns], train["Survived"], test_size=0.1, random_state=i, shuffle=True, stratify =  train["Survived"])
        train_pool = Pool(X_train, y_train, cat_features = categorical_cols)
        val_pool = Pool(X_val, y_val, cat_features = categorical_cols)

        model = CatBoostClassifier(iterations = 1000, loss_function="Logloss", eval_metric="Accuracy", random_seed = i, **params)
        model.fit(train_pool, eval_set = val_pool, early_stopping_rounds = 400, verbose=100)
        predictions = train["Survived"].values - model.predict(train[columns], prediction_type="Class").ravel()
        test_predictions = model.predict(test[columns], prediction_type="Class")
        
        train = pd.concat([train, pd.DataFrame(predictions, columns = ["stack_" + str(i)])], axis=1)
        test = pd.concat([test, pd.DataFrame(test_predictions, columns = ["stack_" + str(i)])], axis=1)
    else:
        X_train, X_val, y_train, y_val = train_test_split(train[columns], train["stack_" + str(i-1)], test_size=0.1, random_state=i, shuffle=True)
        train_pool = Pool(X_train, y_train, cat_features = categorical_cols)
        val_pool = Pool(X_val, y_val, cat_features = categorical_cols)
        model = CatBoostRegressor(iterations = 1000, loss_function="MAE",random_seed = i, **params)
        model.fit(train_pool, eval_set=val_pool, early_stopping_rounds = 400, verbose=100)
        predictions = train["Survived"].values - model.predict(train[columns], prediction_type="RawFormulaVal").ravel()
        test_predictions = test["stack_0"].values - model.predict(test[columns], prediction_type="RawFormulaVal").ravel()
        train = pd.concat([train, pd.DataFrame(predictions, columns = ["stack_" + str(i)])], axis=1)
        test = pd.concat([test, pd.DataFrame(test_predictions, columns = ["stack_" + str(i)])], axis=1)
    
 
# meta_feat = np.column_stack(tuple(pred))
# cv_pool = Pool(train.drop("Survived", axis=1), train["Survived"], cat_features = categorical_cols)
# grid = {
#     "max_leaves": [2**3-1, 2**5-1, 2**7-1],
#     "depth": [6, 7, 8, 9, 10],
#     "l2_leaf_reg": [0, 1, 1.5, 2],
#     "bootstrap_type": ["Bayesian", "Bernoulli", "Poisson"],
#     "min_data_in_leaf": [1, 5, 7, 10, 20, 50],
#     "rsm": [0, 0.1, 0.2, 0.5, 0.7, 1]
# }

# model = CatBoostClassifier(iterations = 1000, loss_function="Logloss", custom_metric="Accuracy", eval_metric="Accuracy")
# model.randomized_search(grid, cv_pool, cv=5, n_iter = 300, verbose=False)
```
