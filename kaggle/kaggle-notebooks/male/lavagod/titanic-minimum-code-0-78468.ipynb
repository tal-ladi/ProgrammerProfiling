- `pandas` for data processing and analysis
- `KNNImputer` imputation for completing missing values using k-Nearest Neighbors
- `DecisionTreeClassifier` a decision tree classifier

```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.tree import DecisionTreeClassifier
```

- `train` training data set
- `test` test data set
- `passenger_id` save passenger indexes of the training dataset

```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
passenger_id = test['PassengerId']
```

- delete columns `Ticket, PassengerId, Name, Cabin` from training data set
- delete columns `Ticket, PassengerId, Name, Cabin` from test data set

```python
train.drop(columns=['Ticket', 'PassengerId', 'Name', 'Cabin'], axis=1, inplace=True)
test.drop(columns=['Ticket', 'PassengerId', 'Name', 'Cabin'], axis=1, inplace=True)
```

- convert categorical variable `Sex, Embarked` into dummy/indicator variables from training data set
- convert categorical variable `Sex, Embarked` into dummy/indicator variables from test data set

```python
train = pd.get_dummies(train, columns=['Sex'])
train = pd.get_dummies(train, columns=['Embarked'])
test = pd.get_dummies(test, columns=['Sex'])
test = pd.get_dummies(test, columns=['Embarked'])
```

- `X` all columns of the training dataset, except `Survived`
- `y` column of the training dataset `Survived`
- `titanic` combining all data from both datasets, except for the column `Survived`

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
titanic = pd.concat([X, test])
```

- `imp` imputation for completing missing values using k-Nearest Neighbors
- to train the model on the most complete data set

```python
imp = KNNImputer(n_neighbors=5)
imp.fit(titanic)
```

- fill in the gaps of the dataset `X`
- fill in the gaps of the dataset `test`

```python
X = imp.transform(X)
test = imp.transform(test)
```

- `model` create a decision tree classifier
- train the model

```python
model = DecisionTreeClassifier(random_state=2343, criterion='entropy', max_depth=6, min_samples_split=5)
model.fit(X, y)
```

- create a response dataset from two columns: 
    - passenger ids
    - model predictions about their survival
- save the result to a file

```python
res = pd.DataFrame({
    'PassengerId': passenger_id,
    'Survived': model.predict(test)
})
res.to_csv('submission.csv', index=False)
```
