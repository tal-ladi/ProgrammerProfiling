```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Load the train and test datasets
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
# Split the train dataset into features and target
X_train = train.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
y_train = train['Survived']
```

```python
# Split the train dataset into features and target
X_train = train.drop('Survived', axis=1)
y_train = train['Survived']
```

```python
# Create an imputation transformer to handle missing values in numerical columns
numerical_imputer = SimpleImputer(strategy='median')
```

```python
# Create a one-hot encoding transformer to handle categorical columns
categorical_transformer = OneHotEncoder()
```

```python
# Create the Random Forest classifier
classifier = RandomForestClassifier(n_estimators=100)
```

```python
# Use the ColumnTransformer class to apply the different transformers to different subsets of the features
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_imputer, ['Age', 'SibSp', 'Parch', 'Fare']),
        ('cat', categorical_transformer, ['Sex', 'Embarked']),
    ])
```

```python
# Create the pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier)
])
```

```python
# Fit the pipeline on the train dataset
pipeline.fit(X_train, y_train)
```

```python
# Make predictions on the test dataset
y_pred = pipeline.predict(test)
```

```python
# Save the predictions to a CSV file
submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': y_pred})
submission.to_csv('submission.csv', index=False)
```
