```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

# Load the train and test datasets
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
test_df = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
# Preprocessing: Encode categorical features
le = LabelEncoder()
train_df['Sex'] = le.fit_transform(train_df['Sex'])
test_df['Sex'] = le.transform(test_df['Sex'])
train_df['Embarked'] = le.fit_transform(train_df['Embarked'])
test_df['Embarked'] = le.transform(test_df['Embarked'])
```

```python
# Prepare the data for modeling
train_df = train_df.fillna(train_df.mean(numeric_only=True))
test_df = test_df.fillna(test_df.mean(numeric_only=True))
```

```python
# Split the training dataset into training and validation sets
X = train_df.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
y = train_df['Survived']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Train the SVM model
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
# Predict on the validation set
y_val_pred = clf.predict(X_val)
```

```python
# Calculate the score
acc = accuracy_score(y_val, y_val_pred)
print("Accuracy: ", acc)
```

```python
# Predict on the test set
test_data = test_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
y_test_pred = clf.predict(test_data)
```

```python
# Store the predictions in a new DataFrame and write it to a CSV file
output = pd.DataFrame({'PassengerId': test_df['PassengerId'], 'Survived': y_test_pred})
output.to_csv('submission.csv', index=False)
```
