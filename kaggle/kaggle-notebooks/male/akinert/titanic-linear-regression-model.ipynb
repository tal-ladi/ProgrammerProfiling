Comment:
* Regression model does not work as we have NaN values in the dataset so i select columns with nonmissing values as features.
* Convert categorical variable into dummy/indicator variables. Such as Men =1 Female = 0 if the passanger is a man.
* 'Sex_female','Pclass_3','Embarked_Q' columns are dropped. For ex. sex_male = 1 have the same meaning sex_female=0 so we dont need sex_female column.

```python
Y_train = train_data["Survived"]
features = ["Fare","Pclass", "Sex", "SibSp", "Parch", "Embarked"]
X_train = pd.get_dummies(train_data[features]) #Convert categorical variable into dummy/indicator variables.
X_train.drop(columns=['Sex_female','Pclass_3','Embarked_Q'], inplace = True)
X_test = pd.get_dummies(test_data[features]) #Convert categorical variable into dummy/indicator variables.
X_test.drop(columns=['Sex_female','Pclass_3','Embarked_Q'], inplace = True)
X_test["Fare"].iloc[152] = test_data["Fare"].mean() # filled 1 null value with mean value of Fare because linear model does not work with missing values
X_test.head()
```

* Values in Fare column has larger amplitude comparing to other columns so i standardized all the columns.

```python
from sklearn.model_selection import train_test_split
X_trainv, X_valid, Y_trainv, Y_valid = train_test_split(X_train, Y_train, test_size=0.2, random_state=5)
print("X_trainv_size =", X_trainv.shape, '\n', "X_valid_size =", X_valid.shape)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(X_trainv)
X_trainvStd = scaler.transform(X_trainv)
X_validStd = scaler.transform(X_valid) #array
pd.DataFrame(X_trainvStd, columns=X_valid.columns).head()
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_trainvStd,Y_trainv)
```

**Since linear model's output is continuous data between 0 and 1, I took above a treshold value as survived(=1) and below as died(=0). (Converted to 0 or 1 by lm.predict(X_train)>i).astype('int'))**

```python
# Calculate the accuracy values for threshold values in range (0.25,0.75) and choose best treshold value
from sklearn.metrics import accuracy_score
score = {}
for i in np.linspace(0.25,0.75,50): #loop over treshold values from 0.25 to 0.75 to select the best treshold 
    score[i] = accuracy_score(Y_valid, (lm.predict(X_validStd)>i).astype('int'))
treshold = max(score, key=score.get) 
print("Max score =",max(score.values()))
print("If predicted value =>",treshold ," passanger survives(=1)")
```

```python
X_testStd = scaler.transform(X_test)
```

```python
# Prediction of test_data
predictions = (lm.predict(X_testStd)>=treshold).astype('int')
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

**In this model "Age" feature was not taken into account because linear regression model does not work with missing values and there are missing values in "Age" column.**
