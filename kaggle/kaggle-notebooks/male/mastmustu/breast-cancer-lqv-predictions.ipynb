```python
df = pd.read_csv(r'/kaggle/input/breast-cancer-wisconsin-data/data.csv')
vectors = pd.read_csv(r'/kaggle/input/bc-lqv-vectors/bc_vectors.csv')
```

```python
vectors = vectors.drop('Unnamed: 0' , axis = 1)
```

```python
X = df.drop('diagnosis' , axis = 1)
X = X.drop('id' , axis = 1)
X = X.drop('Unnamed: 32' , axis = 1)
y = df['diagnosis'].tolist()
```

```python
only_vector = vectors.drop('Tag', axis = 1)
only_vector.shape
```

```python
only_vector_orig = only_vector.copy()
print(only_vector_orig.shape)
```

```python
from scipy.spatial.distance import pdist, squareform
import time 
pred_list = []

start_time = time.time()

for i in range(len(X)) :
    a = X.iloc[i,:].tolist()
    
    only_vector = only_vector_orig.copy()
    only_vector.loc[only_vector.shape[0]] =a 
   
    distances = pdist(only_vector, metric='euclidean')
    dist_matrix = squareform(distances)
    
    b = np.array(dist_matrix[only_vector.shape[0] -1])[:-1]
    
    vector_number = b.argmin()
    
    prediction  = vectors.iloc[vector_number]['Tag']
    pred_list.append(prediction)
    actual = y[i]
    
end_time = time.time()
```

```python
diff = (end_time -start_time)*1000
records = len(pred_list)

print('Processing Time per Transaction is ' ,  str(round(diff/records, 5)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y, pred_list)
```
