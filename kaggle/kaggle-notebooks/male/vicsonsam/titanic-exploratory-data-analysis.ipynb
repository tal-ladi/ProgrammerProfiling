Evaluation Metrics Used 📏
</h2>
</div>

### Based on the problem, we will use the Accuracy metric, perfect for classification.

$Accuracy = \frac {Number of Correct Predictions}{Total Number of Predictions Made} = \frac{TP + TN}{ TP + FP + TN + FN}$

TP: True Positives

FP: False Positivies

TN: True Negatives

FN: False Negatives

#### If you want to learn more about which metrics to use in your Machine Learning Problem, visit:

```python
train_data['Cabin'].fillna('Z', inplace=True)
test_data['Cabin'].fillna('Z', inplace=True)

def cabin_to_list(data):
    cabin_list = []
    for c in data:
        cabin_list.append(c[0])
    return cabin_list

train_data['CabinSection'] = cabin_to_list(train_data['Cabin'])
test_data['CabinSection'] = cabin_to_list(test_data['Cabin'])
```
