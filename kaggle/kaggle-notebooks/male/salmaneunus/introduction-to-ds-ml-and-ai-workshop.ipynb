# Data Preprocessing  - or preparing the dataset for training


```python
# Import label encoder 
from sklearn import preprocessing 
 
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
iris['Species']= label_encoder.fit_transform(iris['Species']) 
  
iris['Species'].unique()
```

```python
iris.Species.dtypes
```

```python
iris.tail(20)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(iris, test_size = 0.25)
print(train.shape)
print(test.shape)
```

**Standardize features by removing the mean and scaling to unit variance**

```python
#Scaling data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

sc = StandardScaler()
sc.fit(train)
train_X_std = sc.transform(train)
test_X_std = sc.transform(test)
```

```python
train_X = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',
                 'PetalWidthCm']]
train_y = train.Species

test_X = test[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',
                 'PetalWidthCm']]
test_y = test.Species
```

```python
train_X.head(10)
```

```python
test_y.head()
```

# Decison Tree Classifier

*Decision Trees (DTs) are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.*

# Define the Models

# Define, Train and test the model

**SVM is a supervised machine learning algorithm which can be used for classification or regression problems. It uses a technique called the kernel trick to transform your data and then based on these transformations it finds an optimal boundary between the possible outputs. Simply put, it does some extremely complex data transformations, then figures out how to seperate your data based on the labels or outputs you've defined.**

```python
#Applying SVC (Support Vector Classification)
from sklearn.svm import SVC

svm = SVC(kernel='rbf', random_state=0, gamma=.10, C=1.0)
svm.fit(train_X_std, train_y)
print('The accuracy of the SVM classifier on training data is {:.2f}'.format(svm.score(train_X_std, train_y)))
print('The accuracy of the SVM classifier on test data is {:.2f}'.format(svm.score(test_X_std,test_y)))
```

*K-nearest neighbors (KNN) algorithm is a type of supervised ML algorithm which can be used for both classification as well as regression predictive problems. However, it is mainly used for classification predictive problems in industry.*

```python
#Applying Knn
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7, p = 2, metric='minkowski')
knn.fit(train_X_std, train_y)

print('The accuracy of the Knn classifier on training data is {:.2f}'.format(knn.score(train_X_std, train_y)))
print('The accuracy of the Knn classifier on test data is {:.2f}'.format(knn.score(test_X_std, test_y)))
```

**Gradient boosting is one of the most powerful techniques for building predictive models. Gradient boosting classifiers are a group of machine learning algorithms that combine many weak learning models together to create a strong predictive model. Decision trees are usually used when doing gradient boosting. Gradient boosting models are becoming popular because of their effectiveness at classifying complex datasets, and have recently been used to win many Kaggle data science competitions.**

```python
#Applying XGBoost
import xgboost as xgb

xgb_clf = xgb.XGBClassifier()
xgb_clf = xgb_clf.fit(train_X_std, train_y)

print('The accuracy of the XGBoost classifier on training data is {:.2f}'.format(xgb_clf.score(train_X_std, train_y)))
print('The accuracy of the XGBoost classifier on test data is {:.2f}'.format(xgb_clf.score(test_X_std,test_y)))
```

```python
#Applying Decision Tree
from sklearn import tree

#Create tree object
decision_tree = tree.DecisionTreeClassifier(criterion='gini')

#Train DT based on scaled training set
decision_tree.fit(train_X_std, train_y)

#Print performance
print('The accuracy of the Decision Tree classifier on training data is {:.2f}'.format(decision_tree.score(train_X_std, train_y)))
print('The accuracy of the Decision Tree classifier on test data is {:.2f}'.format(decision_tree.score(test_X_std, test_y)))
```

**Random forest is a supervised learning algorithm. The "forest" it builds, is an ensemble of decision trees, usually trained with the “bagging” method. The general idea of the bagging method is that a combination of learning models increases the overall result.
Actually random forest builds multiple decision trees and merges them together to get a more accurate and stable prediction.**

```python
#Applying RandomForest
from sklearn.ensemble import RandomForestClassifier

#Create Random Forest object
random_forest = RandomForestClassifier()

#Train model
random_forest.fit(train_X_std, train_y)

#Print performance
print('The accuracy of the Random Forest classifier on training data is {:.2f}'.format(random_forest.score(train_X_std, train_y)))
print('The accuracy of the Random Forest classifier on test data is {:.2f}'.format(random_forest.score(test_X_std,test_y)))
```
