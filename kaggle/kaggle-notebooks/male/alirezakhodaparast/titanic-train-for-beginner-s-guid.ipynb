```python
import pandas as pd 
```

```python
df = pd.read_csv('../input/titanic/train.csv')
df.head()
```

```python
df.shape
```

**Cleaning Data**

```python
df = df.fillna(method = 'pad')
df = df.drop(columns = ['PassengerId','Name','Ticket','Cabin'])
df
```

**Convert Categorical Features or feature with string Value to Numerical Value,e.g: Sex and Embarked feature LabelEncoder Method**

```python
from sklearn.preprocessing import LabelEncoder
```

**Create a model from LabelEncoder and fit_transform method such as le**

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

**Target is Survived feature**

```python
targets = df['Survived']
df = df.drop(columns = ['Survived'])
df
```

```python
from sklearn.model_selection import train_test_split
```

**x_train is your Data**
**y_trani is your Labels or targets**

```python
x_train,x_test,y_train,y_test = train_test_split(df,targets,test_size = 0.3)
```

**SVM = Support Vector Machine**

```python
from sklearn import svm
```

**SVC = Support Vector Classification**

```python
svm_a = svm.SVC()
svm_a.fit(x_train,y_train)
```

**predict x_test data andobtain the accuracy of the SVM algorithm**

```python
from sklearn import metrics
predicted = svm_a.predict(x_test)
print(metrics.classification_report(y_test,predicted))
```

# accuracy of SVM algorithm is 0.67

**Train with decision tree algorithm**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

**predict x_test data andobtain the accuracy of the DecisionTreeClassifier algorithm**

```python
predicted1 = dt.predict(x_test)
print(metrics.classification_report(y_test,predicted1))
```

# accuracy of DecissionTree algorithm is 0.74

# Result: Dession Tree is better accuracy than SVM
