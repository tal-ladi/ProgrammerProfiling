```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split  # split data to train and test
from sklearn.compose import ColumnTransformer 
from sklearn.impute import SimpleImputer   # To remove null values
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('../input/titanic/train.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Let's Plan

```python
df.drop(columns = ['PassengerId', 'Name','Ticket','Cabin'], inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['Survived']),
                                                   df['Survived'],
                                                   test_size = 0.2,
                                                   random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# Imputation transform 
# fills empty cell with mean age

trf1 = ColumnTransformer([
    ('impute_age', SimpleImputer(),[2]),  
    ('impute_embarked',SimpleImputer(strategy = 'most_frequent'), [6])
], remainder = 'passthrough')
```

```python
# one hot encoding

trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse = False, handle_unknown='ignore'),[1,6])
], remainder = 'passthrough')
```

```python
# Scaling

trf3 = ColumnTransformer([
    ('scale', MinMaxScaler(), slice(0,10))
])
```

```python
# Feature Selection
trf4 = SelectKBest(score_func=chi2 , k =8)
```

```python
# Train The Model
trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Pipeline Vs make_pipeline

Pipeline requires naming of steps, make_pipeline does not.

(Same applies to ColumnTransformer vs make_column_transformer)

```python
# Alternate Syntax
#pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```

# Explore the Pipeline

```python
# Details
pipe.named_steps
```

```python
pipe.fit(X_train,y_train)
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross Validation using Pipeline

```python
# cross Validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe , X_train,y_train, cv = 5, scoring = 'accuracy').mean() 
# does train_test_split 5 times as cv = 5 and then removes their accuracy and takes their mean
```

# GridSearch using Pipeline

```python
# Hyper tuning Parameter -- To increase accuracy
# double underscore
params = {
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```
