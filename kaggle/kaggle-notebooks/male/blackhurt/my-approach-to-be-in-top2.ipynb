```python
from sklearn.model_selection import  train_test_split,RepeatedStratifiedKFold,cross_val_score
```

```python
train=pd.read_csv('/kaggle/input/titanic/train.csv',index_col='PassengerId')
test=pd.read_csv('/kaggle/input/titanic/test.csv')
submit=pd.DataFrame(test['PassengerId'])

x=train.drop(['Survived'],axis=1)
y=train['Survived']


xtrain,xvalid,ytrain,yvalid=train_test_split(x,y,test_size=0.2)
```

```python
xtrain['Sex']=xtrain['Sex'].replace({'male':1,'female':0})
xvalid['Sex']=xvalid['Sex'].replace({'male':1,'female':0})
test['Sex']=test['Sex'].replace({'male':1,'female':0})
#------------------------------------------------------------

xtrain['Embarked']=xtrain['Embarked'].replace({'S':0,'C':1,'Q':2})
xvalid['Embarked']=xvalid['Embarked'].replace({'S':0,'C':1,'Q':2})
test['Embarked']=test['Embarked'].replace({'S':0,'C':1,'Q':2})
mmo = {897:1,899:1,930:1,1143:1,1152:1,1153:1,1171:1,972:0,1130:0,1138:0,1173:0,1284:0}
#------------------------------------------------------------------
```

```python
b=[]
for i in range(len(xtrain['Name'])):
    b.append(xtrain['Name'].iloc[i].split(',')[1].split('.')[0])
xtrain['Name']=b
xtrain['Name']=xtrain['Name'].replace([' Dr',' Mlle',' Rev',' Major',' Col',' Don',' the Countess',' Lady',' Jonkheer',' Sir',' Mme',' Ms',' Capt',' Dona'],'Rare')
xtrain['Name']=xtrain['Name'].replace({' Mr':1,' Miss':2,' Mrs':2,' Master':3,'Rare':4})


b=[]
for i in range(len(xvalid['Name'])):
    b.append(xvalid['Name'].iloc[i].split(',')[1].split('.')[0])
xvalid['Name']=b
xvalid['Name']=xvalid['Name'].replace([' Dr',' Mlle',' Rev',' Major',' Col',' Don',' the Countess',' Lady',' Jonkheer',' Sir',' Mme',' Ms',' Capt',' Dona'],'Rare')
xvalid['Name']=xvalid['Name'].replace({' Mr':1,' Miss':2,' Mrs':2,' Master':3,'Rare':4})


b=[]
for i in range(len(test['Name'])):
    b.append(test['Name'].iloc[i].split(',')[1].split('.')[0])
test['Name']=b
test['Name']=test['Name'].replace([' Dr',' Mlle',' Rev',' Major',' Col',' Don',' the Countess',' Lady',' Jonkheer',' Sir',' Mme',' Ms',' Capt',' Dona'],'Rare')
test['Name']=test['Name'].replace({' Mr':1,' Miss':2,' Mrs':2,' Master':3,'Rare':4})
```

```python
for i in range(0,len(xtrain['Cabin'])):
    if(pd.notnull(xtrain['Cabin'].iloc[i])):
        xtrain['Cabin'].iloc[i]=xtrain['Cabin'].iloc[i][0]

        
for i in range(0,len(xvalid['Cabin'])):
    if(pd.notnull(xvalid['Cabin'].iloc[i])):
        xvalid['Cabin'].iloc[i]=xvalid['Cabin'].iloc[i][0]
    
    
for i in range(0,len(test['Cabin'])):
    if(pd.notnull(test['Cabin'].iloc[i])):
        test['Cabin'].iloc[i]=test['Cabin'].iloc[i][0]
```

```python
xtrain['Cabin']=xtrain['Cabin'].replace({'C':1,'B':2,'D':3,'E':4,'F':5,'A':6,'G':7,'T':7})
test['Cabin']=test['Cabin'].replace({'C':1,'B':2,'D':3,'E':4,'F':5,'A':6,'G':7,'T':7})
xvalid['Cabin']=xvalid['Cabin'].replace({'C':1,'B':2,'D':3,'E':4,'F':5,'A':6,'G':7,'T':7})
```

```python
xtrain.drop('Ticket',axis=1,inplace=True)
xvalid.drop('Ticket',axis=1,inplace=True)
test.drop(['Ticket','PassengerId'],axis=1,inplace=True)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier

impute=IterativeImputer(imputation_order='random',max_iter=21)
pipeline=Pipeline(steps=[('a',impute),('s',VarianceThreshold()),('mm',GradientBoostingClassifier())])

pipeline.fit(xtrain,ytrain)

result=pipeline.predict(test)
submit['Survived']=result
submit['Survived'] = submit['Survived'].apply(lambda x: 1 if x>0.8 else 0)
submit['Survived'] = submit.apply(lambda r: mmo[int(r['PassengerId'])] if int(r['PassengerId']) in mmo else r['Survived'], axis=1)
```

```python
submit.to_csv('ver1.csv',index=False)
```
