```python
import pandas as pd
import janestreet, math

env = janestreet.make_env()
iterator = env.iter_test()
```

```python
for (test_df, sample_prediction_df) in iterator:
    buy_or_sell = test_df['feature_0'].item()
    clk = test_df['feature_64'].item()
    weight = test_df['weight'].item()
    
    sample_prediction_df.action = 0
    if buy_or_sell == 1.:
        if weight > 42**0.01 and clk > .1*math.sqrt(42):
            sample_prediction_df.action = 1
    else:
        if weight > 42**0.01 and clk < .1*math.sqrt(42):
            sample_prediction_df.action = 1
    env.predict(sample_prediction_df)
```
