```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor

train= TabularDataset('../input/house-prices-advanced-regression-techniques/train.csv')
test = TabularDataset('../input/house-prices-advanced-regression-techniques/test.csv')

label = 'SalePrice'
time_limit = 60
predictor = TabularPredictor(label=label).fit(train, time_limit=time_limit)

submission = pd.read_csv('../input/house-prices-advanced-regression-techniques/sample_submission.csv')
submission[label] = predictor.predict(test)
submission.to_csv('submission.csv', index=False)
submission.head()
```
