### The average age of those who died in the accident

```python
df1.loc[df1["Survived"] == 0]["Age"].mean()
```

### The average fare of those who died in the accident

```python
df1.loc[df1["Survived"] == 0]["Fare"].mean()
```

### The median of the fare of those who died in the accident

```python
df1.loc[df1["Survived"] == 0]["Fare"].median()
```

### The average age of men who died in the accident

```python
df1.loc[(df1["Survived"] == 0) & (df1["Sex"] == "male")]["Age"].mean()
```

### The average age of women who died in the accident

```python
df1.loc[(df1["Survived"] == 0) & (df1["Sex"] == "female")]["Age"].mean()
```

### The average age of those who survived in the accident

```python
df1.loc[df1["Survived"] == 1]["Age"].mean()
```

### The average fare of those who survived in the accident

```python
df1.loc[df1["Survived"] == 1]["Fare"].mean()
```

### The total number of those who survived in the accident

```python
len(df1.loc[df1["Survived"] == 1])
```

### The median of the fare of those whose age under 10

```python
df1.loc[df1["Age"] < 10]["Fare"].median()
```

### The average fare and the median of those who has first,second and third class ticket

```python
df1.loc[df1["Pclass"]]["Fare"].mean()
```

```python
df1.loc[df1["Pclass"]]["Fare"].median()
```

### The ratio of men who died in the accident

```python
len(df1.loc[(df1["Survived"] == 0) & (df1["Sex"] == "male")]) / len(df1.loc[df1["Sex"] == "male"])
```

### The ratio of women who died in the accident

```python
len(df1.loc[(df1["Survived"] == 0) & (df1["Sex"] == "female")]) / len(df1.loc[df1["Sex"] == "female"])
```
