Design patterns play a crucial role in **software development**, including **data science projects**. They are **reusable solutions to commonly occurring problems** and provide a structured approach to designing and implementing software systems. While design patterns are traditionally associated with object-oriented programming, their principles can be applied to data science projects as well.

Here are some reasons why using design patterns in data science projects is important:

* **Structured Approach**: Data science projects involve various components, such as data extraction, preprocessing, modeling, evaluation, and deployment. Design patterns offer a structured approach to tackle these different stages, ensuring that each component is designed and implemented consistently and efficiently.

* **Reusability**: Design patterns promote code reusability, which is crucial in data science. By encapsulating common functionalities into reusable patterns, you can save time and effort by leveraging existing solutions. This also enhances collaboration among team members as they can easily understand and work with well-known design patterns.

* **Maintainability**: Data science projects often involve complex data pipelines and models. Design patterns provide clear and modular architectures that make the codebase more maintainable. When changes or updates are required, it is easier to locate and modify specific components without impacting the entire system.

* **Scalability**: As data science projects grow in complexity and volume, scalability becomes a critical consideration. Design patterns, such as the Factory pattern or the Strategy pattern, enable the system to adapt to changing requirements and handle larger datasets without significant code modifications.

* **Separation of Concerns**: Data science projects involve multiple concerns, such as data access, feature engineering, model training, and visualization. Design patterns, like the Model-View-Controller (MVC) pattern or the Pipeline pattern, promote separation of concerns, allowing you to focus on each aspect independently. This separation enhances code organization and improves maintainability.

* **Best Practices**: Design patterns embody well-established best practices and design principles. By using design patterns, you leverage the collective experience of software engineers and data scientists who have encountered similar problems in the past. This helps you avoid common pitfalls and adopt proven solutions.

* **Code Understandability**: Design patterns provide a common language and structure that facilitates code understandability among team members. When working on collaborative data science projects, using design patterns can help everyone involved to better comprehend the system architecture and design choices.

# 01. Creational - Factory Method

## 1. Problem:
In software development, you often encounter situations where you need to create objects without specifying their concrete classes. This can arise when you want to decouple the client code from the specific object creation logic or when you need to support multiple object variations or families. The problem is that the client code becomes tightly coupled to the concrete classes, making it difficult to extend or change the object creation process.

## 2. Solution:
The Factory Method design pattern provides a solution by defining an interface or abstract class for creating objects, but delegates the actual object creation to subclasses. Instead of the client code directly instantiating the objects, it relies on the factory method defined in the interface or abstract class. Each subclass can implement the factory method to create the desired objects, encapsulating the object creation logic and allowing for flexibility and extensibility.

## 3. Why it works:
The Factory Method design pattern works because it adheres to the "open-closed principle" - the client code is open for extension (by adding new subclasses) but closed for modification (as it doesn't need to change when new objects are introduced). It allows the client code to depend on abstractions rather than concrete classes, promoting loose coupling and flexibility in object creation.

## Advantages:

* Flexibility: The Factory Method pattern allows for the addition of new object types or variations without modifying the existing client code. New subclasses can be created to handle specific object creation scenarios, providing flexibility in extending the application.
* Encapsulation: The object creation logic is encapsulated within the factory method, making it easier to manage and modify. It separates the responsibility of object creation from the client code, resulting in cleaner and more maintainable code.
* Dependency Inversion: By depending on abstractions rather than concrete classes, the Factory Method pattern supports the principle of dependency inversion. This enables loose coupling and facilitates testing, as dependencies can be easily mocked or replaced.

## Disadvantages:

* Complexity: Implementing the Factory Method pattern introduces additional complexity to the codebase. It requires defining and managing interfaces or abstract classes, as well as creating and maintaining subclasses for each object variation.
* Increased number of classes: The Factory Method pattern may lead to an increased number of classes in the codebase, especially if there are multiple object variations. This can make the codebase larger and potentially harder to understand.


Remember that the advantages and disadvantages may vary depending on the specific context and requirements of your project. It's important to evaluate whether the Factory Method pattern is appropriate for your use case before applying it.

## Python example:

```python
class FeaturePreprocessor(ABC):
    """
    Abstract class for feature preprocessing strategies.
    """

    @abstractmethod
    def preprocess(self, data):
        """
        Abstract method to preprocess the data.
        """
        pass

class AgePreprocessor(FeaturePreprocessor):
    """
    Concrete class for preprocessing the 'Age' feature.
    """

    def preprocess(self, data):
        """
        Preprocesses the 'Age' feature by filling missing values with the median age.
        """
        data['Age'] = data['Age'].fillna(data['Age'].median())
        return data

class FarePreprocessor(FeaturePreprocessor):
    """
    Concrete class for preprocessing the 'Fare' feature.
    """

    def preprocess(self, data):
        """
        Preprocesses the 'Fare' feature by filling missing values with the median fare.
        """
        data['Fare'] = data['Fare'].fillna(data['Fare'].median())
        return data

class PreprocessingFactory:
    """
    Factory class for creating feature preprocessing strategies.
    """

    @staticmethod
    def create_preprocessor(feature_name):
        """
        Factory method to create the appropriate feature preprocessor based on the feature name.
        """
        if feature_name == 'Age':
            return AgePreprocessor()
        elif feature_name == 'Fare':
            return FarePreprocessor()
        else:
            raise ValueError(f"Unsupported feature: {feature_name}")

```

# 02. Creational - Abstract Factory

## 01. Problem:
In software development, you often encounter situations where you need to create families of related objects that share a common theme or have dependencies among them. However, you want to ensure that the created objects are compatible and follow a consistent implementation across the family. The problem is that directly instantiating these objects in client code can result in tight coupling and difficulty in managing object creation and variations.

## 02. Solution:
The Abstract Factory design pattern provides a solution by introducing an abstract factory interface or class that defines methods for creating families of related objects. Each concrete subclass of the abstract factory represents a specific variant or family of objects. The client code interacts with the abstract factory and uses it to create instances of the related objects. This decouples the client code from the concrete implementations and allows for the creation of compatible and consistent object families.

## 03. Why it works:
The Abstract Factory design pattern works because it promotes the principle of dependency inversion and abstraction. The client code depends on the abstract factory interface or class, which allows it to work with any concrete factory subclass without being tightly coupled to specific object types. By using the abstract factory, the client code can create and use objects from a specific family without needing to know the details of their creation or the specific classes involved.

## 04. Advantages:

* **Encapsulation of Object Creation**: The Abstract Factory pattern encapsulates the creation logic within the factory classes, making it easier to manage and modify. It centralizes the object creation process and abstracts it away from the client code, leading to cleaner and more maintainable code.
* **Consistency and Compatibility**: The Abstract Factory pattern ensures that the created objects are compatible and follow a consistent implementation across the entire family. It enforces the creation of related objects that work together seamlessly, promoting proper design and avoiding incompatible object combinations.
* **Flexibility and Extensibility**: The Abstract Factory pattern allows for easy introduction of new families of objects by adding new concrete factory subclasses. This makes it simple to extend the application to support additional object variations or families without modifying the existing client code.

## 05. Disadvantages:

* **Complexity**: Implementing the Abstract Factory pattern can introduce additional complexity to the codebase. It requires defining and managing the abstract factory interface or class, as well as creating and maintaining multiple concrete factory subclasses for each object family. This increased complexity can make the code harder to understand and maintain.
* **Overhead**: The Abstract Factory pattern introduces an additional layer of abstraction, which may result in a performance overhead compared to directly instantiating objects. However, this overhead is generally negligible and often outweighed by the benefits of decoupling and maintainability.

## Python example

```python
class FeatureEngineeringStrategy(ABC):
    """
    Abstract class for feature engineering strategies.
    """

    @abstractmethod
    def engineer_feature(self, data):
        """
        Abstract method to engineer the feature.
        """
        pass

class CategoricalFeatureStrategy(FeatureEngineeringStrategy):
    """
    Concrete class for feature engineering strategy for categorical features.
    """

    def engineer_feature(self, data):
        """
        Engineer categorical features by one-hot encoding the 'Embarked' feature.
        """
        data = pd.get_dummies(data, columns=['Embarked'])
        return data

class NumericFeatureStrategy(FeatureEngineeringStrategy):
    """
    Concrete class for feature engineering strategy for numeric features.
    """

    def engineer_feature(self, data):
        """
        Engineer numeric features by filling missing values with the median.
        """
        data = data.fillna(data.median())
        return data

class FeatureEngineeringFactory:
    """
    Abstract factory class for creating feature engineering strategies.
    """

    @staticmethod
    def create_feature_strategy(feature_type):
        """
        Factory method to create the appropriate feature engineering strategy based on the feature type.
        """
        if feature_type == 'Categorical':
            return CategoricalFeatureStrategy()
        elif feature_type == 'Numeric':
            return NumericFeatureStrategy()
        else:
            raise ValueError(f"Unsupported feature type: {feature_type}")
```

# 03. Creational - Builder

## 01. Problem:
In software development, you often encounter situations where you need to construct complex objects step by step or with varying configurations. The problem is that the traditional approach of directly constructing objects using multiple constructor parameters or setters can become cumbersome, especially when dealing with objects with numerous optional parameters or complex initialization logic.

## 02. Solution:
The Builder design pattern provides a solution by separating the construction of an object from its representation. It involves creating a separate builder class that handles the step-by-step construction of the object. The builder class provides methods for setting the different properties of the object and a method for retrieving the final constructed object. The client code interacts with the builder class to specify the desired configurations and create the complex object.

## 03. Why it works:
The Builder design pattern works by simplifying the object construction process and providing a clear and readable API for constructing complex objects. By separating the construction logic into a builder class, it allows for a more fluent and intuitive way of configuring and creating objects. It also provides a consistent and controlled way of creating objects with various configurations, without the need for a large number of constructors or complex initialization code.

## 04. Advantages:

* **Flexible Object Construction**: The Builder pattern allows for flexible construction of complex objects by providing a step-by-step approach to configure and create objects. It enables the creation of objects with different configurations while maintaining a clear and readable code structure.
* **Encapsulation of Construction Logic**: The construction logic is encapsulated within the builder class, making it easier to manage and modify. The client code interacts with the builder's methods to set the desired properties, while the construction process is hidden from the client code.
* **Readability and Maintainability**: The Builder pattern improves the readability and maintainability of code by providing a more intuitive and descriptive way of constructing objects. It abstracts away the complexity of object creation, making the client code more focused on the desired configurations.

## 05. Disadvantages:

* **Increased Code Complexity**: Implementing the Builder pattern introduces an additional layer of abstraction, which can increase code complexity. It requires defining the builder class and its associated methods, which may add extra code overhead compared to direct object instantiation.
* **Overhead in Small Objects**: The Builder pattern may introduce unnecessary overhead for simple objects that do not have many optional parameters or complex initialization logic. In such cases, the simplicity and readability of direct object instantiation may be more appropriate.

## Python example

```python
class DataPreprocessor:
    """
    Class representing the data preprocessing pipeline.
    """

    def __init__(self):
        self.steps = []

    def add_step(self, step):
        """
        Adds a preprocessing step to the pipeline.
        """
        self.steps.append(step)

    def preprocess_data(self, data):
        """
        Applies the preprocessing steps to the data in sequential order.
        """
        for step in self.steps:
            data = step.execute(data)
        return data

class PreprocessingStep(ABC):
    """
    Abstract class representing a preprocessing step.
    """

    @abstractmethod
    def execute(self, data):
        """
        Abstract method to execute the preprocessing step.
        """
        pass

class FillMissingValuesStep(PreprocessingStep):
    """
    Concrete class representing a preprocessing step to fill missing values.
    """

    def execute(self, data):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        data['Age'] = data['Age'].fillna(data['Age'].median())
        data['Fare'] = data['Fare'].fillna(data['Fare'].median())
        return data

class FeatureEngineeringStep(PreprocessingStep):
    """
    Concrete class representing a preprocessing step for feature engineering.
    """

    def execute(self, data):
        """
        Performs feature engineering operations on the data.
        """
        # Example: Creating a new feature 'FamilySize'
        data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
        return data
```

# 04. Creational - Prototype

## 01. Problem:
In software development, you often encounter situations where you need to create new objects by copying existing objects. However, creating objects from scratch can be expensive in terms of resources and time, especially when the initialization process is complex or involves costly operations. The problem is that directly creating objects using constructors or factories can lead to inefficient object creation and reduced performance.

## 02. Solution:
The Prototype design pattern provides a solution by introducing a prototype object that serves as a blueprint for creating new objects. Instead of creating objects from scratch, you create a clone of the prototype object and modify it as necessary. The clone operation can be shallow or deep, depending on the complexity of the object and its internal references. This approach allows you to efficiently create new objects by copying existing ones, reducing the overhead of object creation.

## 03. Why it works:
The Prototype design pattern works by leveraging object cloning to create new objects. By using a prototype object as a blueprint, you eliminate the need for costly initialization or construction operations. Instead, you simply clone the prototype and modify the clone if required. This approach allows for efficient object creation and customization while maintaining a consistent and optimized object creation process.

## 04. Advantages:

* **Reduced Object Creation Cost**: The Prototype pattern reduces the cost of creating new objects by avoiding expensive initialization or construction operations. It leverages object cloning, which is often more efficient and faster than creating objects from scratch.
* **Customizable Object Creation**: With the Prototype pattern, you can easily customize and modify the cloned objects as needed. You can modify specific attributes or relationships of the cloned objects to tailor them to your requirements without the need for complex configuration or instantiation steps.
* **Flexibility and Extensibility**: The Prototype pattern provides flexibility and extensibility in object creation. You can introduce new prototypes or subclasses to create variations of objects, and the cloning process can be adjusted to support different cloning strategies, such as shallow or deep cloning.
Disadvantages:

* **Cloning Overhead**: The Prototype pattern introduces the overhead of cloning objects, which may be negligible for simple objects but can become noticeable for complex or large objects. Deep cloning, in particular, can have performance implications if the object contains deep nested structures or references.
* **Complexity in Managing Prototypes**: Managing and maintaining the prototypes can become complex, especially when dealing with multiple prototype objects or complex inheritance hierarchies. Ensuring proper initialization and updating of prototypes can require additional effort and careful management.

## Python example

```python
class DataPreprocessingConfig:
    """
    Class representing the data preprocessing configuration.
    """

    def __init__(self):
        self.fill_missing_values = True
        self.feature_scaling = False
        self.feature_selection = False

    def configure(self, fill_missing_values=True, feature_scaling=False, feature_selection=False):
        """
        Configures the data preprocessing settings.
        """
        self.fill_missing_values = fill_missing_values
        self.feature_scaling = feature_scaling
        self.feature_selection = feature_selection

    def clone(self):
        """
        Creates a clone of the configuration object.
        """
        return deepcopy(self)

class DataPreprocessor:
    """
    Class representing the data preprocessor.
    """

    def __init__(self, config):
        self.config = config

    def preprocess_data(self, data):
        """
        Applies the data preprocessing steps based on the configuration.
        """
        if self.config.fill_missing_values:
            # Perform fill missing values preprocessing
            data = self.fill_missing_values(data)

        if self.config.feature_scaling:
            # Perform feature scaling preprocessing
            data = self.feature_scaling(data)

        if self.config.feature_selection:
            # Perform feature selection preprocessing
            data = self.feature_selection(data)

        return data

    def fill_missing_values(self, data):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        # Example: Filling missing values in 'Age' column with median
        data['Age'] = data['Age'].fillna(data['Age'].median())
        return data

    def feature_scaling(self, data):
        """
        Applies feature scaling to selected numeric columns.
        """
        # Example: Applying Min-Max scaling to 'Fare' column
        data['Fare'] = (data['Fare'] - data['Fare'].min()) / (data['Fare'].max() - data['Fare'].min())
        return data

    def feature_selection(self, data):
        """
        Applies feature selection based on selected criteria.
        """
        # Example: Removing 'Cabin' column
        data = data.drop('Cabin', axis=1)
        return data
```

# 05. Creational - Singleton

## 01. Problem:
In software development, you often encounter situations where you need to ensure that a class has only one instance throughout the program. This can be necessary for resource management, coordination, or to maintain a central point of access for shared functionality. The problem is that multiple instances of the class can be created, leading to unnecessary resource allocation or inconsistent behavior.

## 02. Solution:
The Singleton design pattern provides a solution by ensuring that a class has only one instance and providing global access to that instance. It involves defining a class with a private constructor and a static method that controls the creation and access to the instance. The first time the static method is called, it creates the instance and returns it. On subsequent calls, it simply returns the existing instance.

## 03. Why it works:
The Singleton design pattern works by enforcing the creation of a single instance of the class and providing a global point of access to that instance. By restricting the constructor to be private, it prevents external code from directly creating multiple instances. The static method ensures that only one instance is created and returned, regardless of how many times it is called.

## 04. Advantages:

* **Single Instance**: The Singleton pattern ensures that there is only one instance of the class throughout the program, eliminating the possibility of multiple instances being created and providing a global point of access to that instance.
* **Global Access:** The Singleton pattern provides a globally accessible point to access the instance, making it convenient to use and reducing the need for passing references to the instance across different parts of the code.
Resource Management: The Singleton pattern can be useful for managing resources that should be shared across the program, such as database connections, file handles, or hardware devices.


## 05. Disadvantages:

* **Global State:** The Singleton pattern introduces global state, which can make it harder to track dependencies and understand the flow of data within the program. This can increase the complexity of debugging and testing.
* **Reduced Flexibility**: The Singleton pattern limits the extensibility and flexibility of the class, as it allows only one instance to exist. This can make it difficult to substitute the singleton object with a different implementation or mock it for testing purposes.
* **Testing Challenges**: Testing code that relies on the Singleton pattern can be challenging, as the global state introduced by the singleton instance can lead to dependencies and interference between test cases.

## Python example

```python
class DataPreprocessingConfig:
    """
    Class representing the data preprocessing configuration.
    """

    def __init__(self):
        self.fill_missing_values = True
        self.feature_scaling = True
        self.feature_selection = False

class DataPreprocessingManager:
    """
    Singleton class for managing the data preprocessing configuration.
    """

    _instance = None

    @staticmethod
    def get_instance():
        """
        Static method to get the singleton instance.
        """
        if not DataPreprocessingManager._instance:
            DataPreprocessingManager._instance = DataPreprocessingConfig()
        return DataPreprocessingManager._instance

    def preprocess_data(self, data):
        """
        Applies the data preprocessing steps based on the configuration.
        """
        config = self.get_instance()
        if config.fill_missing_values:
            # Perform fill missing values preprocessing
            data = self.fill_missing_values(data)

        if config.feature_scaling:
            # Perform feature scaling preprocessing
            data = self.feature_scaling(data)

        if config.feature_selection:
            # Perform feature selection preprocessing
            data = self.feature_selection(data)

        return data

    def fill_missing_values(self, data):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        # Example: Filling missing values in 'Age' column with median
        data['Age'] = data['Age'].fillna(data['Age'].median())
        return data

    def feature_scaling(self, data):
        """
        Applies feature scaling to selected numeric columns.
        """
        # Example: Applying Min-Max scaling to 'Fare' column
        data['Fare'] = (data['Fare'] - data['Fare'].min()) / (data['Fare'].max() - data['Fare'].min())
        return data

    def feature_selection(self, data):
        """
        Applies feature selection based on selected criteria.
        """
        # Example: Removing 'Cabin' column
        data = data.drop('Cabin', axis=1)
        return data

```

# 06. Structural - Adapter

## 01. Problem:
In software development, you often encounter situations where you need to integrate or use existing classes or components that have incompatible interfaces or communication protocols. The problem is that direct communication or interaction with these incompatible classes is not possible, resulting in the inability to reuse or integrate them seamlessly.

## 02. Solution:
The Adapter design pattern provides a solution by acting as a bridge between the incompatible interfaces or protocols. It involves creating an adapter class that wraps around the incompatible class and provides a compatible interface that other components can interact with. The adapter class translates requests or calls from the compatible interface into the corresponding calls or operations of the underlying incompatible class.

## 03. Why it works:
The Adapter design pattern works by introducing an intermediary adapter class that acts as a layer of translation or compatibility. It allows components that require a specific interface or protocol to communicate with the incompatible class without needing to modify the existing codebase or the incompatible class itself. The adapter class encapsulates the translation logic, providing a seamless integration point for the incompatible components.

## 04. Advantages:

* **Improved Reusability**: The Adapter pattern allows for the reuse of existing incompatible classes or components by providing a compatible interface. It enables integration with other components or systems without the need for significant modifications to the existing codebase.
* **Seamless Integration**: The Adapter pattern provides a seamless integration point between the incompatible components. It acts as a bridge, translating and adapting requests and responses between the components, ensuring smooth communication.
* **Encapsulation of Compatibility Logic**: The adapter class encapsulates the logic required to translate or adapt the incompatible interface or protocol. This encapsulation helps maintain clean and modular code, as the translation logic is isolated within the adapter.

## 05. Disadvantages:

* **Increased Complexity**: Implementing the Adapter pattern can introduce additional complexity, as it requires creating and managing the adapter class and its interactions with the incompatible class. This added complexity can make the code harder to understand and maintain.
* **Performance Overhead**: The Adapter pattern may introduce a performance overhead due to the translation or adaptation process. The additional layer of abstraction and the need to map between different interfaces or protocols can impact the overall performance of the system.

# 07. Structural - Bridge

## 01. Problem:
In software development, you often encounter situations where you need to decouple an abstraction from its implementation. The problem is that a traditional approach with a single class hierarchy for both the abstraction and its implementation can result in a large number of class combinations and a tight coupling between the abstraction and implementation, making the codebase complex and hard to maintain.

## 02. Solution:
The Bridge design pattern provides a solution by separating the abstraction from its implementation using two separate class hierarchies. It involves creating an abstraction class that defines the high-level interface, and an implementation class that provides the concrete implementation details. The abstraction class contains a reference to the implementation class, allowing the abstraction to delegate specific implementation tasks to the implementation class.

## 03. Why it works:
The Bridge design pattern works by decoupling the abstraction from its implementation, allowing them to vary independently. By encapsulating each hierarchy within its own class, the Bridge pattern promotes loose coupling and flexibility. The abstraction can select and use different implementations at runtime without modifying its code, and new abstractions and implementations can be added independently without affecting each other.

## 04. Advantages:

* **Decoupling of Abstraction and Implementation**: The Bridge pattern separates the abstraction and implementation, reducing the tight coupling between them. This separation promotes modularity, flexibility, and maintainability, as changes in one hierarchy do not affect the other.
* **Runtime Selection of Implementations**: The Bridge pattern allows for the dynamic selection of different implementations for the abstraction at runtime. This provides the flexibility to switch implementations without modifying the abstraction's code, making it easier to adapt to changing requirements.
* **Extensibility**: The Bridge pattern allows for the independent extension of the abstraction and implementation hierarchies. New abstractions or implementations can be added without affecting existing classes, providing a scalable and modular design.

## 05. Disadvantages:

* **Increased Complexity**: Implementing the Bridge pattern introduces additional complexity, as it requires defining and managing two separate class hierarchies. This can make the codebase more intricate and harder to understand, especially for smaller or simpler projects.
* **Overhead of Abstraction and Implementation Interaction**: The Bridge pattern introduces an additional layer of interaction between the abstraction and implementation classes, which can incur some performance overhead. This overhead may be negligible in most cases, but it should be considered in performance-critical scenarios.

## Python example

```python
class DataPreprocessor(ABC):
    """
    Abstraction class representing the data preprocessing.
    """

    def __init__(self, implementation):
        self.implementation = implementation

    @abstractmethod
    def preprocess_data(self, data):
        """
        Abstract method to preprocess the data.
        """
        pass

class FillMissingValues(DataPreprocessor):
    """
    Concrete implementation class for filling missing values in the data.
    """

    def preprocess_data(self, data):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        return self.implementation.fill_missing_values(data)

class FeatureScaling(DataPreprocessor):
    """
    Concrete implementation class for performing feature scaling on the data.
    """

    def preprocess_data(self, data):
        """
        Applies feature scaling to selected numeric columns.
        """
        return self.implementation.feature_scaling(data)

class TitanicDataPreprocessor:
    """
    Implementation class providing specific data preprocessing strategies for the Titanic dataset.
    """

    def fill_missing_values(self, data):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        # Example: Filling missing values in 'Age' column with median
        data['Age'] = data['Age'].fillna(data['Age'].median())
        return data

    def feature_scaling(self, data):
        """
        Applies feature scaling to selected numeric columns.
        """
        # Example: Applying Min-Max scaling to 'Fare' column
        data['Fare'] = (data['Fare'] - data['Fare'].min()) / (data['Fare'].max() - data['Fare'].min())
        return data

```

# 08. Structural - Composite

## 01. Problem:
In software development, you often encounter situations where you need to work with hierarchical structures or collections of objects in a uniform way. The problem is that treating individual objects and groups of objects differently can lead to complex and inconsistent code. Additionally, the client code may need to perform operations on both individual objects and groups of objects without knowing their specific types.

## 02. Solution:
The Composite design pattern provides a solution by treating individual objects and groups of objects uniformly. It involves creating a common interface for both individual objects and groups of objects, allowing clients to interact with them in a consistent way. The individual objects and groups of objects are structured as a tree-like hierarchy, where both share a common base class or interface.

## 03. Why it works:
The Composite design pattern works by representing individual objects and groups of objects in a unified manner through a common interface or base class. This abstraction allows clients to treat objects uniformly without needing to know their specific types. The hierarchical structure of objects enables recursive operations, where operations can be applied to both individual objects and groups of objects seamlessly.

## 04. Advantages:

* **Transparent Structure**: The Composite pattern provides a transparent structure for working with hierarchies of objects. Clients can treat individual objects and groups of objects uniformly, abstracting away the details of the object hierarchy.
* **Flexibility and Scalability**: The Composite pattern allows for the addition of new types of objects without impacting the existing code. New individual objects and groups of objects can be easily added to the hierarchy, providing flexibility and scalability.
* **Simplified Client Code**: Clients can work with the objects in a consistent and unified manner, without needing to handle different object types separately. This simplifies client code and promotes a more intuitive and clean design.

## 05. Disadvantages:

* **Complexity**: Implementing the Composite pattern can introduce additional complexity, especially when dealing with complex object hierarchies or operations. The recursive nature of the pattern may require careful handling to avoid performance issues or unintended behaviors.
* **Limited Type-Specific Operations**: Treating objects uniformly through a common interface may restrict the usage of type-specific operations. Clients may need to perform type checks or casts to access specific features or behaviors of individual objects.
* **Maintenance Overhead**: Managing the relationships and hierarchy between objects can require additional effort and maintenance. Ensuring consistency and integrity within the object hierarchy may become more challenging as the complexity of the structure increases.

## Python example


```python
import pandas as pd

class PassengerComponent:
    """
    Abstract base class representing the component in the composite structure.
    """

    def get_passenger_count(self):
        """
        Abstract method to get the passenger count.
        """
        pass

class Passenger(PassengerComponent):
    """
    Leaf node representing an individual passenger in the composite structure.
    """

    def __init__(self, name):
        self.name = name

    def get_passenger_count(self):
        """
        Returns the passenger count, which is always 1 for an individual passenger.
        """
        return 1

class PassengerGroup(PassengerComponent):
    """
    Composite node representing a group of passengers in the composite structure.
    """

    def __init__(self, name):
        self.name = name
        self.passengers = []

    def add_passenger(self, passenger):
        """
        Adds a passenger to the group.
        """
        self.passengers.append(passenger)

    def remove_passenger(self, passenger):
        """
        Removes a passenger from the group.
        """
        self.passengers.remove(passenger)

    def get_passenger_count(self):
        """
        Returns the passenger count, which is the sum of the passenger counts in the group.
        """
        count = 0
        for passenger in self.passengers:
            count += passenger.get_passenger_count()
        return count

```

# 09. Structural - Decorator

## 01. Problem:

The problem is that traditional inheritance-based approaches can result in a rigid class hierarchy or introduce a high number of subclasses for each combination of functionalities, making the codebase complex and hard to maintain.

## 02. Solution:
The Decorator design pattern provides a solution by allowing you to add or modify the behavior of an object at runtime without affecting its structure. It involves creating decorator classes that wrap around the original object and provide additional functionalities by implementing the same interface as the original object. Multiple decorators can be stacked on top of each other, forming a chain of decorators.

## 03. Why it works:
The Decorator design pattern works by extending the functionality of an object dynamically. It works by creating decorator classes that implement the same interface as the original object and contain a reference to the original object. The decorators add new functionalities by delegating the calls to the original object and adding their own behavior before or after the delegation, effectively extending or modifying the behavior of the original object.

## 04. Advantages:

* **Flexible Extension**: The Decorator pattern allows for the dynamic addition of functionalities to an object without modifying its structure or code. Decorators can be easily added, removed, or combined, providing flexible extension capabilities.
* **Open-Closed Principle**: The Decorator pattern follows the Open-Closed Principle, as it allows for extending the behavior of an object without modifying its existing code. New decorators can be added without altering the existing classes or their interfaces.
* **Composability**: Decorators can be stacked on top of each other, forming a chain of decorators. This composability allows for fine-grained control over the added functionalities and enables the creation of complex combinations of behaviors.

## 05.Disadvantages:

* **Increased Complexity**: Implementing the Decorator pattern can introduce additional complexity, especially when dealing with a large number of decorators or complex interaction between decorators. The chain of decorators can become intricate, and the order of their execution may affect the final behavior.
* **Potential Performance Impact**: The use of decorators can introduce a slight performance overhead due to the added indirection and the potential for multiple method calls in the chain of decorators. However, this overhead is typically negligible unless used extensively in performance-critical sections.

## Python example:

```python
class DataPreprocessor:
    """
    Component class representing the data preprocessing pipeline.
    """

    def preprocess(self, data):
        """
        Preprocesses the data.
        """
        raise NotImplementedError

class MissingValueFiller(DataPreprocessor):
    """
    Concrete decorator class for filling missing values in the data.
    """

    def __init__(self, component):
        self.component = component

    def preprocess(self, data):
        """
        Performs missing value filling on the data.
        """
        # Fill missing values in selected columns with appropriate strategies
        data = self.fill_missing_values(data)

        # Call the preprocess method of the component
        if self.component:
            data = self.component.preprocess(data)

        return data

    def fill_missing_values(self, data):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        # Example: Filling missing values in 'Age' column with median
        data['Age'] = data['Age'].fillna(data['Age'].median())
        return data

class FeatureScaler(DataPreprocessor):
    """
    Concrete decorator class for performing feature scaling on the data.
    """

    def __init__(self, component):
        self.component = component

    def preprocess(self, data):
        """
        Performs feature scaling on the data.
        """
        # Apply feature scaling to selected numeric columns
        data = self.feature_scaling(data)

        # Call the preprocess method of the component
        if self.component:
            data = self.component.preprocess(data)

        return data

    def feature_scaling(self, data):
        """
        Applies feature scaling to selected numeric columns.
        """
        # Example: Applying Min-Max scaling to 'Fare' column
        data['Fare'] = (data['Fare'] - data['Fare'].min()) / (data['Fare'].max() - data['Fare'].min())
        return data

```

# 10. Structural - Facade

## 01. Problem:
In software development, you often encounter situations where a complex subsystem or set of classes with various dependencies needs to be simplified and presented with a unified, simplified interface. The problem is that interacting directly with the individual components of the subsystem can be cumbersome, requiring deep knowledge of the system's internal structure and dependencies.

## 02. Solution:
The Facade design pattern provides a solution by introducing a simplified interface or facade that hides the complexities of the underlying subsystem. It involves creating a facade class that encapsulates the interactions with the subsystem's components and provides a single entry point for clients to access the subsystem's functionalities. The facade class coordinates the actions and interactions between the subsystem's components, shielding the clients from the details of the subsystem.

## 03. Why it works:
The Facade design pattern works by providing a high-level interface that hides the complexities of the subsystem. It encapsulates the interactions and dependencies between the subsystem's components, allowing clients to interact with the subsystem through a simplified, unified interface. By centralizing the interactions in the facade class, it becomes easier to understand, use, and maintain the subsystem, as clients can work with a consistent and simplified API.

## 04. Advantages:

* **Simplified Interface**: The Facade pattern provides a simplified and unified interface to a complex subsystem. Clients can interact with the subsystem through the facade class without needing to understand the internal structure or dependencies of the subsystem's components.
* **Decoupling**: The Facade pattern decouples the subsystem from its clients by introducing an intermediary facade class. Changes in the subsystem's internal structure or dependencies are isolated within the facade class, minimizing the impact on the client code.
* **Improved Readability and Maintainability**: The Facade pattern improves the readability and maintainability of the codebase by providing a clear, high-level API. It abstracts away the complexities of the subsystem, making the codebase more organized and easier to understand.

## 05. Disadvantages:

* **Limited Flexibility**: The Facade pattern may limit the flexibility or customization options available to clients. The simplified interface provided by the facade class may not expose all the functionalities or configurations of the subsystem's components.
* **Increased Indirection**: The introduction of the facade class adds an extra layer of indirection between clients and the subsystem. This can lead to a small performance overhead due to the additional method calls and abstraction.

## Python example

```python
class DataPreprocessorFacade:
    """
    Facade class providing a simplified interface for data preprocessing tasks.
    """

    def __init__(self, data):
        self.data = data

    def preprocess_data(self):
        """
        Performs common data preprocessing tasks.
        """
        self.remove_duplicates()
        self.fill_missing_values()
        self.normalize_data()
        self.encode_categorical_features()

        # Additional preprocessing steps can be added here

    def remove_duplicates(self):
        """
        Removes duplicate rows from the dataset.
        """
        self.data.drop_duplicates(inplace=True)

    def fill_missing_values(self):
        """
        Fills missing values in selected columns with appropriate strategies.
        """
        # Example: Filling missing values in 'Age' column with median
        self.data['Age'] = self.data['Age'].fillna(self.data['Age'].median())

    def normalize_data(self):
        """
        Normalizes selected numeric columns in the dataset.
        """
        # Example: Normalizing 'Fare' column
        self.data['Fare'] = (self.data['Fare'] - self.data['Fare'].mean()) / self.data['Fare'].std()

    def encode_categorical_features(self):
        """
        Encodes categorical features in the dataset.
        """
        # Example: One-hot encoding 'Embarked' column
        encoded_embarked = pd.get_dummies(self.data['Embarked'], prefix='Embarked')
        self.data = pd.concat([self.data, encoded_embarked], axis=1)
        self.data.drop('Embarked', axis=1, inplace=True)

```

# 11. Structural - Flyweight

## 01. Problem:
In software development, you often encounter situations where you need to efficiently represent a large number of objects with similar properties or intrinsic data. The problem is that representing each object individually can consume significant memory and resources, especially when many objects share common state or have overlapping characteristics.

## 02. Solution:
The Flyweight design pattern provides a solution by separating the intrinsic and extrinsic states of objects. It involves creating a flyweight object that represents the shared state among multiple objects. The extrinsic state, which is unique to each object, is stored externally. The flyweight object is shared among multiple objects, reducing memory consumption and improving performance.

## 03. Why it works:
The Flyweight design pattern works by promoting object sharing and reusing. By separating the intrinsic and extrinsic states, the flyweight object can be shared among multiple objects, effectively reducing the memory footprint. The extrinsic state is managed externally and passed to the flyweight objects when needed, enabling multiple objects to reuse the shared state efficiently.

## 04. Advantages:

* **Memory Efficiency**: The Flyweight pattern reduces memory consumption by sharing common state among multiple objects. This is especially useful when dealing with a large number of objects or when memory usage is a concern.
* **Performance Improvement**: By sharing common state, the Flyweight pattern can improve performance by reducing the need for object creation and memory allocation. It can also minimize the overhead associated with redundant state storage and manipulation.
* **Enhanced Scalability**: The Flyweight pattern promotes scalability by allowing the system to handle a large number of objects efficiently. It allows for dynamic addition or removal of flyweight objects without affecting the behavior of other objects.

## 05. Disadvantages:

* **Increased Complexity**: Implementing the Flyweight pattern can introduce additional complexity, as it requires distinguishing between intrinsic and extrinsic states and managing the interaction between them. This complexity can make the codebase harder to understand and maintain, especially for smaller or simpler projects.
* **State Sharing Limitation**: The Flyweight pattern is most effective when the objects have significant common state to share. If the objects have very distinct or unique states, the benefits of sharing may be limited, and the overhead of managing the flyweight objects may outweigh the advantages.

## Python example

```python
class CategoryFlyweight:
    """
    Flyweight class representing a category.
    """

    def __init__(self, category):
        self.category = category
        
class CategoricalDataFlyweightFactory:
    """
    Flyweight factory class managing the flyweight objects.
    """

    def __init__(self):
        self.categories = {}

    def get_category(self, category):
        """
        Retrieves the flyweight object for the given category.
        If it doesn't exist, creates a new one.
        """
        if category not in self.categories:
            self.categories[category] = CategoryFlyweight(category)
        return self.categories[category]

class TitanicDataProcessor:
    """
    Data processor class utilizing the Flyweight pattern to optimize memory usage for categorical data.
    """

    def __init__(self):
        self.flyweight_factory = CategoricalDataFlyweightFactory()

    def process_data(self, data):
        """
        Processes the Titanic dataset by optimizing memory usage for categorical data.
        """
        processed_data = data.copy()

        # Convert categorical columns to flyweight objects
        categorical_columns = ['Pclass', 'Sex', 'Embarked']
        for column in categorical_columns:
            processed_data[column] = processed_data[column].apply(self.get_flyweight_category)

        return processed_data

    def get_flyweight_category(self, category):
        """
        Retrieves the flyweight object for the given category.
        """
        return self.flyweight_factory.get_category(category)

```

# 12. Structural - Proxy

## 01. Problem:
In software development, you often encounter situations where you need to control access to an object or provide additional functionality without modifying its underlying implementation. The problem is that direct access to the object may not be appropriate or desired due to security concerns, resource limitations, or the need for additional operations before or after accessing the object.

## 02. Solution:
The Proxy design pattern provides a solution by introducing a proxy object that acts as a surrogate or placeholder for the real object. The proxy object controls access to the real object and can add extra behavior or perform additional operations before or after forwarding the request to the real object. The proxy object mimics the interface of the real object, allowing clients to interact with it in a transparent manner.

## 03. Why it works:
The Proxy design pattern works by introducing an intermediary object between the client and the real object. The proxy object provides a level of indirection, allowing it to control access to the real object and perform additional operations as necessary. Clients interact with the proxy object as if it were the real object, unaware of the underlying mechanisms and modifications made by the proxy.

## 04. Advantages:

* **Access Control**: The Proxy pattern enables fine-grained control over access to the real object. The proxy object can enforce access restrictions or authentication mechanisms, ensuring that clients adhere to defined rules and security requirements.
* **Resource Management**: Proxies can manage the creation and destruction of expensive or limited resources associated with the real object. This allows for efficient resource utilization and prevents unnecessary resource allocation.
* **Additional Functionality**: Proxies can add extra functionality or perform additional operations before or after accessing the real object. This can include caching, logging, lazy initialization, or access optimization, providing enhanced capabilities without modifying the real object.

## 05. Disadvantages:

* **Increased Complexity**: Introducing a proxy object can increase the overall complexity of the system. The proxy needs to replicate the interface of the real object, potentially leading to a larger codebase and more intricate design.
* **Performance Overhead**: The use of a proxy object can introduce a performance overhead due to the additional method calls and indirection. However, the impact is typically minimal unless the proxy is performing complex operations or if the system heavily relies on proxy instances.

## Python example

```python
class TitanicDataset:
    """
    Real object class representing the Titanic dataset.
    """

    def __init__(self):
        self._data = None

    def load_data(self):
        """
        Loads the Titanic dataset from a source (e.g., CSV file).
        """
        if self._data is None:
            # Load the dataset from a source (e.g., CSV file)
            self._data = pd.read_csv('/kaggle/input/titanic/train.csv')

    def get_data(self):
        """
        Returns the Titanic dataset.
        """
        return self._data

class TitanicDatasetProxy:
    """
    Proxy class controlling access to the Titanic dataset and adding caching functionality.
    """

    def __init__(self):
        self._dataset = TitanicDataset()

    def get_data(self):
        """
        Returns the Titanic dataset, loading it if necessary and caching the result.
        """
        if self._dataset.get_data() is None:
            self._dataset.load_data()
        return self._dataset.get_data()

```

# 13. Behavioral - Interpreter

## 01. Problem:
In software development, you often encounter scenarios where you need to interpret or evaluate structured textual expressions or rules. The problem is that implementing a grammar or rule set directly in code can be complex, inflexible, and difficult to maintain, especially when the grammar or rule set is large or subject to frequent changes.

## 02. Solution:
The Interpreter design pattern provides a solution by defining a language, grammar, or set of rules using a combination of classes and objects. It involves creating interpreter classes that represent the grammar rules or expressions and provide methods for their interpretation or evaluation. These interpreter classes work together to parse, evaluate, and execute the structured expressions.

## 03. Why it works:
The Interpreter design pattern works by separating the grammar or rule set from its implementation. It allows you to define the rules or expressions using a structured representation that can be easily modified or extended. The interpreter classes interpret or evaluate these expressions based on the defined grammar, allowing you to build complex behavior by combining simple interpreters.

## 04. Advantages:

* **Extensibility**: The Interpreter pattern enables easy extension of the grammar or rule set. New rules or expressions can be added by creating new interpreter classes without modifying the existing ones, providing flexibility and maintainability.
* **Separation of Concerns**: The pattern separates the grammar or rule set from its implementation, promoting a clear separation of concerns. The interpreter classes handle the interpretation or evaluation of the expressions, while the client code focuses on utilizing the interpreters to achieve the desired behavior.
* **Readability and Maintainability**: The Interpreter pattern can improve the readability and maintainability of code by providing a structured and modular approach to representing and executing expressions. The grammar rules are defined in a clear and concise manner, making the codebase easier to understand and modify.

## 05. Disadvantages:

* **Complexity**: Implementing the Interpreter pattern can introduce additional complexity, especially when dealing with large or complex grammars. Defining and coordinating multiple interpreter classes may require a good understanding of the grammar and careful design to ensure efficient parsing and interpretation.
* **Performance Overhead**: Depending on the complexity of the grammar and the size of the input expressions, the Interpreter pattern may introduce a performance overhead. The interpretation and evaluation process can involve multiple method calls and object interactions, which may impact the overall performance, especially for computationally intensive tasks.

## Python example

```python

class AgeRuleInterpreter:
    """
    Interpreter class for evaluating age-based selection rules.
    """

    def evaluate(self, data, rule):
        """
        Evaluates the age-based selection rule on the dataset.
        """
        passengers = []
        for index, row in data.iterrows():
            if eval(rule):
                passengers.append(row['Name'])
        return passengers


```

The problem is that directly implementing the entire algorithm in a single class can lead to code duplication, lack of flexibility, and difficulty in managing variations or customization.

## 02. Solution:
The Template Method design pattern provides a solution by defining the skeleton of an algorithm in a base class, while delegating the implementation of specific steps to subclasses. It involves creating an abstract base class that provides a template method defining the overall algorithm structure. The template method calls several abstract or hook methods that subclasses must implement to customize or extend the algorithm's behavior.

## 03. Why it works:
The Template Method design pattern works by separating the high-level algorithm from the specific implementations of its steps. The base class defines the common structure and controls the overall flow of the algorithm through the template method. Subclasses can provide their own implementations for the abstract or hook methods, allowing customization and extension while adhering to the algorithm's structure.

## 04. Advantages:

* **Code Reusability**: The Template Method pattern promotes code reuse by providing a common algorithm structure in the base class. Subclasses can focus on implementing specific steps without duplicating the entire algorithm.
* **Flexibility and Customization**: The pattern allows for flexibility and customization by letting subclasses override or extend specific steps of the algorithm. Subclasses can tailor the behavior to their specific needs without modifying the overall algorithm.
* **Encapsulation**: The Template Method pattern encapsulates the algorithm's structure in the base class, hiding the details of individual steps. This improves code organization, readability, and maintainability.

## 05. Disadvantages:

* **Limited Runtime Flexibility**: Once the algorithm's structure is defined in the base class, it may be challenging to modify or extend the structure dynamically at runtime. Changes to the algorithm's structure typically require modifications to the base class and its subclasses.
* **Potential Tight Coupling**: The Template Method pattern may lead to tight coupling between the base class and its subclasses. Subclasses need to adhere to the structure and contract defined by the base class, which can limit flexibility and make the codebase more coupled.

# 15. Behavioral - Chain of Responsibility


## 01. Problem:
In software development, you often encounter scenarios where multiple objects or components need to handle a request or perform an operation sequentially. The problem is that the handling logic may vary, and the responsibility chain may change dynamically. Hard-coding the handling logic or creating tight coupling between objects can make the codebase inflexible and difficult to maintain.

## 02. Solution:
The Chain of Responsibility design pattern provides a solution by decoupling the sender of a request from its receivers. It involves creating a chain of handler objects, where each handler has the ability to handle the request or pass it to the next handler in the chain. The request traverses the chain until it is handled or reaches the end of the chain.

## 03. Why it works:
The Chain of Responsibility design pattern works by separating the sender of a request from its receivers and allowing multiple objects to have a chance to handle the request. Each handler in the chain can decide whether to handle the request or pass it to the next handler. This promotes loose coupling, flexibility, and scalability, as the chain can be modified at runtime, and new handlers can be added or removed without affecting the client code.

## 04. Advantages:

* **Flexibility and Extensibility**: The Chain of Responsibility pattern allows for flexible handling logic and easy extension of the responsibility chain. New handlers can be added or removed dynamically, and the handling sequence can be altered without modifying the client code.
* **Decoupling**: The pattern decouples the sender of a request from its receivers, reducing the dependencies and allowing for independent evolution of both the sender and the receivers.
* **Reusability**: Handlers in the chain can be reused in different chains or scenarios. They can be combined in various ways to handle different types of requests or operations.

## 05. Disadvantages:

* **Overhead and Performance Impact**: The Chain of Responsibility pattern can introduce an overhead due to the traversal of the chain and the potential for multiple objects to handle the request. This can impact performance, especially if the chain is long or if there are many potential handlers for each request.
* **Potential Unhandled Requests**: If the chain is not properly configured or if there is no appropriate handler for a request, it may remain unhandled, leading to unexpected behavior or errors. Proper care must be taken to ensure that all requests are handled appropriately.

# 16. Behavioral - Command

## 01. Problem:
In software development, you often encounter scenarios where you need to decouple the sender of a request from its receiver, allowing you to parameterize clients with different requests and support undoable operations. The problem is that directly invoking operations on objects can tightly couple the sender and receiver, making it challenging to manage different requests or support undo operations.

## 02. Solution:
The Command design pattern provides a solution by encapsulating a request as an object, decoupling the sender from the receiver. It involves creating command objects that encapsulate a specific operation or request along with its parameters. The command objects provide an execute method that triggers the corresponding operation on the receiver object.

## 03. Why it works:
The Command design pattern works by abstracting the sender and receiver through the command object. The sender does not need to know the details of how the request is executed or who the receiver is. It only needs to hold and execute the command object. This loose coupling enables the sender to work with different command objects, allowing for easy parameterization and dynamic composition of requests.

## 04. Advantages:

* **Decoupling**: The Command pattern decouples the sender from the receiver, promoting a more flexible and maintainable codebase. The sender and receiver can evolve independently without affecting each other.
* **Undo/Redo Support**: Command objects can be designed to support undoable operations. By storing the necessary information and logic, the commands can be reversed or reapplied, providing undo and redo functionality.
* **Extensibility**: The pattern allows for easy extension by adding new commands without modifying the existing client code. This makes it straightforward to incorporate new operations or requests.

## 05. Disadvantages:

* **Increased Complexity**: The Command pattern can introduce additional complexity, especially when dealing with a large number of commands and command variations. Managing the command objects and their interactions may require careful design and coordination.
* **Overhead**: Using the Command pattern may introduce a slight performance overhead due to the additional layer of abstraction and the need to create and manage command objects. However, the impact is usually negligible unless the system heavily relies on command invocations.

# 17. Behavioral - Iterator

## 01. Problem:
In software development, you often need to traverse or iterate over elements in a collection or aggregate object without exposing its underlying structure. The problem is that different collection types may have different internal structures, and client code may need to be modified when iterating over each type of collection. Additionally, you may need to iterate over the elements of a collection in different ways without modifying the collection itself.

## 02. Solution:
The Iterator design pattern provides a solution by decoupling the traversal or iteration logic from the underlying collection or aggregate object. It involves defining an iterator object that encapsulates the iteration behavior and provides methods for accessing the elements of the collection. The iterator object keeps track of the current position within the collection and provides methods for moving to the next element.

## 03. Why it works:
The Iterator design pattern works by separating the iteration logic from the collection or aggregate object. The iterator provides a standardized interface for accessing elements, abstracting away the specific structure or implementation details of the collection. This allows client code to iterate over different collection types using a consistent interface without needing to modify the client code for each collection type.

## 04. Advantages:

* **Decoupling**: The Iterator pattern decouples the client code from the underlying collection structure, allowing the collection to evolve independently without affecting the client code.
* **Abstraction**: The pattern provides an abstraction layer that hides the internal details of the collection or aggregate object. This promotes code simplicity, maintainability, and code reuse.
* **Multiple Iteration Strategies**: The Iterator pattern allows for the definition of multiple iterator objects for the same collection, each with its own traversal strategy. This provides flexibility in iterating over the elements of a collection in different ways without modifying the collection itself.

## 05. Disadvantages:

* **Limited Navigation**: Iterators typically support sequential navigation, moving forward one element at a time. Random access or backward traversal may not be efficiently supported by all iterator implementations.
* **Performance Overhead**: The Iterator pattern may introduce a slight performance overhead due to the additional layer of abstraction and method calls involved in the iteration process. However, the impact is usually negligible unless the collection is very large or the iteration is performed frequently.

```python
class TitanicDatasetIterator:
    """
    Iterator class for iterating over the Titanic dataset.
    """

    def __init__(self, dataset):
        self.dataset = dataset
        self.current_index = 0

    def __iter__(self):
        """
        Returns itself as an iterator.
        """
        return self

    def __next__(self):
        """
        Returns the next row in the dataset.
        """
        if self.current_index >= len(self.dataset):
            raise StopIteration

        row = self.dataset.iloc[self.current_index]
        self.current_index += 1
        return row

```

# 19. Behavioral - Memento

## 01. Problem:
In software development, you often encounter scenarios where you need to capture and restore an object's internal state without violating encapsulation. The problem is that directly exposing the internal state or providing public methods for state manipulation can compromise encapsulation and introduce complexity in managing object state changes.

## 02. Solution:
The Memento design pattern provides a solution by introducing three main components: the originator, the memento, and the caretaker. The originator represents the object whose state needs to be captured and restored. The memento is an object that stores the state of the originator. The caretaker is responsible for managing and storing the mementos. The originator creates mementos to save its state, and the caretaker stores and retrieves mementos as needed.

## 03. Why it works:
The Memento design pattern works by encapsulating and externalizing the object's state through mementos. The originator is responsible for creating mementos and restoring its state from a memento when needed. The caretaker manages the mementos, allowing for the saving and retrieval of object states without violating encapsulation.

## 04. Advantages:

* **Encapsulation**: The Memento pattern encapsulates the internal state of an object within mementos, preserving encapsulation and information hiding.
* **Undo/Redo Support**: The pattern allows for the capture and restoration of object states, enabling undo and redo functionality.
* **State History**: The caretaker can store multiple mementos, providing a history of object states that can be navigated and restored as needed.

## 05. Disadvantages:

* **Increased Memory Usage**: The Memento pattern can consume additional memory, especially when storing multiple mementos to support undo/redo or maintaining a large state history.
* **Performance Overhead**: Creating and restoring mementos can introduce a performance overhead, especially if the state being captured is large or if there are frequent state changes.

# 20. Behavioral - Observer

## 01. Problem:
In software development, you often encounter scenarios where multiple objects need to be notified and updated when the state of another object changes. The problem is that establishing direct dependencies between these objects can lead to tight coupling and code that is difficult to maintain, extend, and reuse.

## 02. Solution:
The Observer design pattern provides a solution by defining a one-to-many dependency between objects, where the subject (or observable) maintains a list of observers (or subscribers) and notifies them automatically when its state changes. The observers can then update themselves based on the new state of the subject.

## 03. Why it works:
The Observer design pattern works by decoupling the subject and observers, allowing them to vary independently. The subject only needs to know about the abstract Observer interface, while observers implement this interface to receive notifications. This loose coupling enables flexible and dynamic communication between objects.

## 04. Advantages:

* **Decoupling**: The Observer pattern decouples the subject and observers, reducing direct dependencies and improving code maintainability, extensibility, and reusability.
* **Flexibility**: New observers can be added or existing ones can be removed without modifying the subject. Similarly, subjects can change without affecting the observers, as long as they adhere to the Observer interface.
* **Event-driven Architecture**: The Observer pattern facilitates event-driven architecture, where objects communicate through events and notifications, enabling loosely coupled and highly modular systems.

## 05. Disadvantages:

* **Unintended Updates**: Observers may receive unnecessary or redundant updates if the subject's state changes frequently or if the update mechanism is not properly managed.
* **Performance Impact**: Notifying a large number of observers can impact performance, especially if observers perform computationally expensive operations or if the subject updates frequently. Care should be taken to optimize the notification process if performance is a concern.

## Python example

# 21. Behavioral - State

## 01. Problem:
In software development, you often encounter scenarios where an object's behavior changes based on its internal state. The problem is that implementing conditional statements or switch-case statements to manage different behaviors can lead to complex and tightly coupled code that is difficult to maintain, extend, and understand.

## 02. Solution:
The State design pattern provides a solution by encapsulating each behavior in a separate state object and allowing the object to dynamically change its behavior by switching between these states. The object delegates the behavior to its current state object, decoupling the behavior from the object itself.

## 03. Why it works:
The State design pattern works by representing each behavior as a separate state class. The object that exhibits the behavior maintains a reference to its current state object and delegates method calls to the state object. When the object's state needs to change, it switches to a different state object. This approach enables the object to have different behaviors without altering its own implementation.

## 04. Advantages:

* **Encapsulated Behavior**: Each state is encapsulated within its own class, promoting better organization and separation of concerns. Each state object can be developed and tested independently.
* **Simplified Object**: The State pattern simplifies the object by removing conditional logic or switch-case statements, resulting in cleaner and more maintainable code.
* **Dynamic Behavior**: The object can dynamically change its behavior by switching between different states at runtime, allowing for flexible and extensible behavior changes.

## 05. Disadvantages:

* **Increased Number of Classes**: Implementing the State pattern may introduce a larger number of classes, especially if there are many different states and complex behavior variations.
* **State Transition Logic**: The responsibility of managing state transitions may become complex, especially if there are complex dependencies or specific rules for switching between states.

## Python example

```python
class PreprocessingState:
    """
    Abstract base class representing a preprocessing state.
    """

    def preprocess(self, data):
        """
        Abstract method to perform preprocessing based on the state.
        """
        raise NotImplementedError("Subclasses must implement preprocess().")

class SimplePreprocessingState(PreprocessingState):
    """
    Concrete subclass representing a simple preprocessing state.
    """

    def preprocess(self, data):
        """
        Preprocesses the data using a simple strategy.
        """
        print("Performing simple preprocessing...")
        # Perform simple preprocessing logic

class AdvancedPreprocessingState(PreprocessingState):
    """
    Concrete subclass representing an advanced preprocessing state.
    """

    def preprocess(self, data):
        """
        Preprocesses the data using an advanced strategy.
        """
        print("Performing advanced preprocessing...")
        # Perform advanced preprocessing logic

class PreprocessingStateMachine:
    """
    State machine class that manages the preprocessing states.
    """

    def __init__(self):
        self.states = {
            "simple": SimplePreprocessingState(),
            "advanced": AdvancedPreprocessingState()
        }
        self.current_state = self.states["simple"]

    def set_state(self, state):
        """
        Sets the current preprocessing state.
        """
        self.current_state = self.states[state]

    def preprocess_data(self, data):
        """
        Preprocesses the data based on the current state.
        """
        self.current_state.preprocess(data)


```

# 23. Behavioral - Visitor

## 01. Problem:
In software development, you often encounter scenarios where you need to perform operations on a group of objects that belong to different classes with different interfaces. The problem is that adding new operations or functionalities to these objects becomes challenging as it requires modifying the existing classes, violating the Open-Closed Principle.

## 02. Solution:
The Visitor design pattern provides a solution by separating the operations from the object structure. It introduces a Visitor class that defines separate visit methods for each type of object in the structure. Each object accepts a visitor and delegates the operation to the visitor, decoupling the objects from the specific operations performed on them.

## 03. Why it works:
The Visitor design pattern works by leveraging double dispatch. The object structure accepts a visitor and calls the appropriate visit method based on its own type. The visitor then performs the desired operation on the object. This approach allows adding new operations or functionalities without modifying the existing object classes.

## 04. Advantages:

* **Extensibility**: The Visitor pattern allows adding new operations to the object structure by introducing new visitor classes, without modifying the existing object classes.
* **Separation of Concerns**: The pattern separates the logic of operations from the object structure, promoting a clean and modular design.
* **Open-Closed Principle**: The Visitor pattern follows the Open-Closed Principle by allowing the addition of new operations without modifying existing code.

## 05. Disadvantages:

* **Object Structure Awareness**: The Visitor pattern requires the object structure to be aware of the visitor and call the appropriate visit method. This may introduce coupling between the object structure and the visitor.
* **Object Hierarchy Impact**: Adding new types of objects to the hierarchy may require modifying the visitor interface and all its implementations, potentially impacting existing code.

## Python example
