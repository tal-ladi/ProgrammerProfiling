```python
import tensorflow as tf
import numpy as np 

model = tf.saved_model.load('/kaggle/input/distilbert-nlp-with-disaster-tweets/DistilBERT')
classes = [  "0" ,  "1" ,  ]
```

```python
Id=[]
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input/nlp-with-disaster-tweets/train'):
    for filename in filenames:
        Id.append(os.path.join(dirname, filename))
Id[:5]
```

```python
train=pd.DataFrame()
train=train.assign(filename=Id)
train.head()
```

```python
train['label']=train['filename']
train['label']=train['label'].str.replace('/kaggle/input/nlp-with-disaster-tweets/train/','')
train.head()
```

```python
train['label'] = train['label'].str.split('/').str[0]
train.head()
```

```python
Id=[]
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input/nlp-with-disaster-tweets/test'):
    for filename in filenames:
        Id.append(os.path.join(dirname, filename))
Id[:5]
```

```python
test=pd.DataFrame()
test=test.assign(filename=Id)
test.head()
```

```python
test['label']=test['filename']
test['label']=test['label'].str.replace('/kaggle/input/nlp-with-disaster-tweets/test/','')
test.head()
```

```python
test['label'] = test['label'].str.split('/').str[0]
test.head()
```

```python
from distilbert_tokeniser import tokenize
result=[]
for i in test.filename:
    sentence = open(i).read()
    tok = tokenize(sentence ,  max_len=32)
    t1 = tf.constant( np.array([tok["input_ids"] ]) , dtype='int32'  )
    class_scores = model(t1)[0].numpy()
    result.append(classes[class_scores.argmax()])
result[:5]
```

```python
test=test.assign(prediction=result)
test.head()
```

```python
sol=pd.read_csv('/kaggle/input/nlp-getting-started/sample_submission.csv')
sol=sol[['id']]
sol.head()
```

```python
sol['filename']='/kaggle/input/nlp-with-disaster-tweets/sol/'+sol['id'].astype('str')+'.txt'
sol.head()
```

```python
result=[]
for i in sol.filename:
    sentence = open(i).read()
    tok = tokenize(sentence ,  max_len=32)
    t1 = tf.constant( np.array([tok["input_ids"] ]) , dtype='int32'  )
    class_scores = model(t1)[0].numpy()
    result.append(classes[class_scores.argmax()])
result[:5]
```

```python
sol=sol.assign(target=result)
sol=sol[['id','target']]
sol.head()
```
