**Data preprocessing** refers to the technique of preparing (cleaning and organizing) the raw data to make it suitable for a building and training Machine Learning models.

**step 1 : Finding and cleaning null values**

```python
training.isna().sum() # for training
```

**<p style="color:red">Observations 📋</p>**

🔘 For Age , cabin , Embarked there are null values.

🔘 Null values in cabin are to many.

**<p style="color:green">solutions </p>**

🔘 Process both age and Embarked because They don't have a lot of nulls, Embarked with mode, age average for each of both sexes separately

🔘 Delete cabin column it has a lot of nulls

```python
fill=training.Embarked.mode()[0]
fill
```

```python
training.Embarked.fillna(fill,inplace=True)
```

```python
training.drop('Cabin',axis=1,inplace=True)
```

```python
male_age_fill=training[training['Sex']=='male']['Age'].mean().astype(int)
female_age_fill=training[training['Sex']=='female']['Age'].mean().astype(int)
male_age_fill,female_age_fill
```

```python
training.loc[(training.Age.isnull()) & (training.Sex=='male'),'Age']=male_age_fill # fill with mean of males to all null males

training.loc[(training.Age.isnull()) & (training.Sex=='female'),'Age']=female_age_fill# fill with mean of females to all null females
```

```python
training.isna().sum()
```

```python
testing.isna().sum() # for testing
```

**<p style="color:red">Observations 📋</p>**

🔘 For Age , cabin ,fare there are null values.

🔘 Null values in cabin are to many.

**<p style="color:green">solutions </p>**

🔘 Process both age  because it don't have a lot of nulls,  average for each of both sexes separately, fare by deleting the row because it's 1 row ,It does not affect the data.

🔘 Delete cabin column it has a lot of nulls.

```python
male_age_fill=testing[testing['Sex']=='male']['Age'].mean().astype(int)
female_age_fill=testing[testing['Sex']=='female']['Age'].mean().astype(int)
male_age_fill,female_age_fill
```

```python
testing.loc[(testing.Age.isnull()) & (testing.Sex=='male'),'Age']=male_age_fill # fill with mean of males to all null males

testing.loc[(testing.Age.isnull()) & (testing.Sex=='female'),'Age']=female_age_fill# fill with mean of females to all null females

testing.drop('Cabin',axis=1,inplace=True)

testing.drop(testing.loc[testing["Fare"].isnull()].index,inplace=True)

testing.isna().sum()
```

🔘 **seperate the title from the name**

```python
training['Title'] =training['Name'].apply(lambda x:x.split(',')[1].split('.')[0].strip())
testing['Title']  =testing['Name'].apply(lambda x:x.split(',')[1].split('.')[0].strip())
```

```python
#befor filtering
training.Title.value_counts()
```

```python
training['Title'] = training['Title'].replace(['Lady', 'the Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer'],'a_little')
training['Title'] = training['Title'].replace(['Mlle','Ms'], 'Miss')
training['Title'] = training['Title'].replace('Mme', 'Mrs')


testing['Title'] = testing['Title'].replace(['Lady', 'the Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer','Dona'],'a_little')
testing['Title'] = testing['Title'].replace(['Mlle','Ms'], 'Miss')
testing['Title'] = testing['Title'].replace('Mme', 'Mrs')
```

🔘**let's encoding the data**



🔘 Models only work with numerical values. For this reason, it is necessary to convert the categorical values of the features into      numerical ones, So the machine can learn from those data and gives the right model. This process of converting categorical data into         numerical data is called Encoding.

🔘 **Label encoder** :Sklearn provides a very efficient tool for encoding the levels of categorical features into numeric values. LabelEncoder encode labels with a value between 0 and n_classes-1 where n is the number of distinct labels. If a label repeats it assigns the same value to as assigned earlier.

**<p style="color:red">Note : </p>**

also we can use one hot encoder or pandas.get_dummies function with diffrent output but finally the same task


```python
training.head()
```

```python
#for training data

to_categorical=['Sex','Embarked','Title']

from sklearn.preprocessing import LabelEncoder

L_encoder=LabelEncoder()

for label in to_categorical:
    training[label]=L_encoder.fit_transform(training[label])


```

```python
#for testing data

to_categorical=['Sex','Embarked','Title']

from sklearn.preprocessing import LabelEncoder

L_encoder=LabelEncoder()

for label in to_categorical:
    testing[label]=L_encoder.fit_transform(testing[label])
```

🔘 **Dividing Fare into 4 category**


```python
training['fare_category']=pd.cut(training['Fare'],4,labels=[1,2,3,4]) #Divide Fare into 4 sections (for training data)
testing['fare_category']=pd.cut(testing['Fare'],4,labels=[1,2,3,4]) #Divide Fare into 4 sections (for testing data)
```

🔘 **Delete useless columns in the data**

```python
training.drop(['PassengerId','Name','Age','SibSp','Parch','Parch','Ticket','Fare'],axis=1,inplace=True)
testing.drop(['Name','Age','SibSp','Parch','Parch','Ticket','Fare'],axis=1,inplace=True)
```

<h1 style="background-color:black;color:white;font-family:newtimeroman;font-size:350%;text-align:center;border-radius: 15px 50px;"> Data splitting 📏  </h1>


```python
x_train = training.drop('Survived',axis=1)
y_train = training['Survived']

x_test  = testing.drop('PassengerId',axis=1)

print('x train shape :',x_train.shape)
print('y train shape :',y_train.shape)
print('x test shape :',x_test.shape)
```

<a id="table"></a>
<h1 style="background-color:black;color:white;font-family:newtimeroman;font-size:350%;text-align:center;border-radius: 15px 50px;"> Model building ⚙️ </h1>


```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()

LR.fit(x_train,y_train)

LR.score(x_train,y_train)*100
```

```python
from sklearn.svm import SVC

svr=SVC()

svr.fit(x_train,y_train)

svr.score(x_train,y_train)*100
```

```python
from sklearn.tree import DecisionTreeClassifier

DT=DecisionTreeClassifier()

DT.fit(x_train,y_train)

DT.score(x_train,y_train)*100
```

```python
from sklearn.ensemble import RandomForestClassifier

RF=RandomForestClassifier()

RF.fit(x_train,y_train)

RF.score(x_train,y_train)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()

knn.fit(x_train,y_train)

knn.score(x_train,y_train)*100
```

**<p style="color:red">Observations 📋</p>**

🔘 It turns out that the best models are Decision Tree Classifier and  Random Forest Classifier
