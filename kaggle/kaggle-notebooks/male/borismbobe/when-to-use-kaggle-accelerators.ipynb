### **Hello and good evening to all readers of this document.**

Today I'm going to talk about an important tool of kaggle: **Accelerators**

When we work on kaggle's jupyter notebook development environment, kaggle gives us hardware resources that are
1. **HDD hard drive**
1. **RAM (Random Access Memory)**
1. **And the GPU(Graphic Processing Unit).** 
who all have limits
The accelerators are there to temporarily increase these resources.

We have two types of kaggle speedups:
1. **GPU(Graphic Processing Unit)**
1. **TPU(Tensor Processing Unit,)**

when to use these accelerators

when you run a python or R script, check on the process bar as in the screenshot:


![](https://www.googleapis.com/download/storage/v1/b/kaggle-forum-message-attachments/o/inbox%2F9655897%2Fd03ee03d91a3aa7bdfe01fb030b13843%2Fimage2.jpg?generation=1660565006918156&alt=media)

If the process bar turns red like in the screenshot, then you should use kaggle accelerators

![](https://www.googleapis.com/download/storage/v1/b/kaggle-forum-message-attachments/o/inbox%2F9655897%2Faeb44328df69cc5aac82251186141d4d%2Fimage1.jpg?generation=1660565154551430&alt=media)
