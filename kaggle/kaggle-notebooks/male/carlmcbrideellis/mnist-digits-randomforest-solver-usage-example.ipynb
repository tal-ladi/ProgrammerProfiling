### [MNIST digits RandomForest solver](https://www.kaggle.com/carlmcbrideellis/mnist-digits-randomforest-solver): Usage example

```python
import numpy  as np
import pandas as pd
import pickle
```

Read in the model. 
The model requires the input to be a numpy array of shape `(1, 784)`

```python
RandomForest_model = pickle.load(open('../input/mnist-digits-randomforest-solver/MNIST_RandomForest_model.sav','rb'))
```

Usage example: Selecting a row from the [Digit Recognizer](https://www.kaggle.com/c/digit-recognizer) test dataset and returning its value:

```python
test_df = pd.read_csv('../input/digit-recognizer/test.csv')
```

```python
row = 27996
image_to_classify = test_df.iloc[row,:].to_numpy().reshape(1, 784)

RandomForest_model.predict(image_to_classify)[0]
```
