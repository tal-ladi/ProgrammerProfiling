```python
import pandas as pd
import datetime
from IPython.display import Markdown
from IPython.display import Image

n_users = pd.read_csv("../input/meta-kaggle/Users.csv", usecols=["Id"]).nunique().item()

# competitions
n_competitions = pd.read_csv("../input/meta-kaggle/Competitions.csv", usecols=["Id"]).nunique().item()
n_submissions = pd.read_csv("../input/meta-kaggle/Submissions.csv", usecols=["Id"]).nunique().item()
n_sub_per_comp = int(n_submissions/n_competitions)
Competitions = pd.read_csv("../input/meta-kaggle/Competitions.csv")
# 2022-12-09: InClass is now called Community
#n_InClass = Competitions.HostSegmentTitle.value_counts()["InClass"]
n_Community = Competitions.HostSegmentTitle.value_counts()["Community"]
n_Featured = Competitions.HostSegmentTitle.value_counts()["Featured"]
n_Research = Competitions.HostSegmentTitle.value_counts()["Research"]
n_Playground = Competitions.HostSegmentTitle.value_counts()["Playground"]
n_comp_with_cash = Competitions["RewardType"].value_counts()["USD"]
total_dollars = int(Competitions.query("RewardType == 'USD' ").RewardQuantity.sum())
Competitions['EnabledDate'] = pd.to_datetime(Competitions['EnabledDate'])
first_comp = Competitions.query("HostSegmentTitle == 'Featured' ").EnabledDate.min()
first_comp_month_year = first_comp.strftime('%B of %Y')

# datasets
n_datasets = pd.read_csv("../input/meta-kaggle/Datasets.csv", usecols=["Id"]).nunique().item()
n_dataset_authors = pd.read_csv("../input/meta-kaggle/Datasets.csv", usecols=["CreatorUserId"]).nunique().item()
datasets_per_author = round(n_datasets/n_dataset_authors, 1)

# notebooks
n_notebooks = pd.read_csv("../input/meta-kaggle/Kernels.csv", usecols=["Id"]).nunique().item()
n_notebook_authors = pd.read_csv("../input/meta-kaggle/Kernels.csv", usecols=["AuthorUserId"]).nunique().item()
notebooks_per_author = round(n_notebooks/n_notebook_authors, 1)

# topics and comments
n_topics = pd.read_csv("../input/meta-kaggle/ForumTopics.csv", usecols=["Id"]).nunique().item()
n_comments = pd.read_csv("../input/meta-kaggle/ForumMessages.csv", usecols=["Id"]).nunique().item()
comments_per_topic = round(n_comments/n_topics, 1)

# Tiers stuff
Users = pd.read_csv("../input/meta-kaggle/Users.csv")
n_GMs = Users.query("PerformanceTier == 4").shape[0]
n_Ms  = Users.query("PerformanceTier == 3").shape[0]
n_Es  = Users.query("PerformanceTier == 2").shape[0]
n_Cs  = Users.query("PerformanceTier == 1").shape[0]
n_Ns  = Users.query("PerformanceTier == 0").shape[0]
n_kaggle_staff  = Users.query("PerformanceTier == 5").shape[0]

# Thanks to the work of @steubk and his dataset:
# https://www.kaggle.com/steubk/meta-kagglemaster-achievements-snapshot
# we now have access to extra data that was in the old "UserAchievements.csv" file
# Note: this file is not updated when Meta Kaggle updates, so there may be slight discrepancies
MasterAchievements = pd.read_csv("../input/meta-kagglemaster-achievements-snapshot/MasterAchievements.csv")
# GM stuff
MasterAchievements['GM_tier_count'] = MasterAchievements[MasterAchievements.astype(str) == 'grandmaster'].count(axis=1)
n_4xGM = MasterAchievements.query("GM_tier_count == 4").shape[0]
n_3xGM = MasterAchievements.query("GM_tier_count == 3").shape[0]
n_2xGM = MasterAchievements.query("GM_tier_count == 2").shape[0]
# M stuff
MasterAchievements['M_tier_count'] = MasterAchievements[MasterAchievements.astype(str) == 'master'].count(axis=1)
n_4xM = MasterAchievements.query("(M_tier_count == 4) and (GM_tier_count == 0)").shape[0]
n_3xM = MasterAchievements.query("(M_tier_count == 3) and (GM_tier_count == 0)").shape[0]
n_2xM = MasterAchievements.query("(M_tier_count == 2) and (GM_tier_count == 0)").shape[0]

# --------------- printing section --------------

site_logo = Image(url='https://www.kaggle.com/static/images/site-logo.png')
display(site_logo)
display(Markdown('### There are ' + f'{n_users:,}' + '  kaggle user accounts.'))
display(Markdown('### The first kaggle competition was launched in ' + first_comp_month_year))
display(Markdown('### There have been  ' + f'{n_competitions:,}' + '  competitions held, with ' 
                                         + f'{n_submissions:,}' + " submissions made. This is an average of "
                                         + f'{n_sub_per_comp:,}' + " submissions per competition."))
display(Markdown('### Of these,  ' + f'{n_Featured:,}' + '  were featured competitions, ' 
                                   + f'{n_Research:,}' + " research competitions, "
                                   + f'{n_Playground:,}' + " playground competitions, and "
                                   + f'{n_Community:,}' + " were community competitions."))
display(Markdown('### There have been  ' + f'{n_comp_with_cash:,}' + ' competitions that awarded cash, with a total of $'
                                         + f'{total_dollars:,}' + ' in prize money.'))
display(Markdown('### There are  ' + f'{n_datasets:,}' + '  public datasets by '
                                   + f'{n_dataset_authors:,}' + ' authors; an average of ' 
                                   + f'{datasets_per_author:,}' + '  datasets per author.'))
display(Markdown('### There are  ' + f'{n_notebooks:,}' + '  public notebooks by '
                                   + f'{n_notebook_authors:,}' + ' authors; an average of ' 
                                   + f'{notebooks_per_author:,}' + '  notebooks per author.'))
display(Markdown('### There have been  ' + f'{n_topics:,}' + '  topics posted, with ' 
                                         + f'{n_comments:,}' + " comments made; an average of " 
                                         + f'{comments_per_topic:,}' + " comments per topic."))

def print_markdown(string, color=None):
    colour_text = "<span style='font-size:17px;font-weight: bold;color:{}'>{}</span>".format(color, string)
    display(Markdown(colour_text))
    
tier_image = Image(url='https://www.kaggle.com/static/images/tiers/grandmaster@48.png')
display(tier_image)
print_markdown('There are  ' + f'{n_GMs:,}' + '  kaggle Grandmasters.',  color="#ddaa17")
print_markdown('Out of these there are ' + f'{n_4xGM:,}' + ' quadruple Grandmasters, '
                                     + f'{n_3xGM:,}' + ' triple Grandmasters, and '
                                     + f'{n_2xGM:,}' + ' double Grandmasters.',  color="#ddaa17")
tier_image = Image(url='https://www.kaggle.com/static/images/tiers/master@48.png')
display(tier_image)
print_markdown('There are  ' + f'{n_Ms:,}' + '  kaggle Masters.',  color="#f76629")
print_markdown('Out of these there are ' + f'{n_4xM:,}' + ' quadruple Masters, '
                                     + f'{n_3xM:,}' + ' triple Masters, and '
                                     + f'{n_2xM:,}' + ' double Masters.',  color="#f76629")
tier_image = Image(url='https://www.kaggle.com/static/images/tiers/expert@48.png')
display(tier_image)
print_markdown('There are  ' + f'{n_Es:,}' + '  kaggle Experts.',  color="#96508e")
tier_image = Image(url='https://www.kaggle.com/static/images/tiers/contributor@48.png')
display(tier_image)
print_markdown('There are  ' + f'{n_Cs:,}' + '  kaggle Contributors.',  color="#20beff")
tier_image = Image(url='https://www.kaggle.com/static/images/tiers/novice@48.png')
display(tier_image)
print_markdown('There are  ' + f'{n_Ns:,}' + '  kaggle Novices.',  color="#4fcb93")
tier_image = Image(url='https://raw.githubusercontent.com/Carl-McBride-Ellis/images_for_kaggle/main/black_circle_48px.png')
display(tier_image)
display(Markdown('### There are  ' + f'{n_kaggle_staff:,}' + '  kaggle staff.'))

# add a blank line as a spacer
display(Markdown(""))

now = datetime.datetime.now()
display(Markdown('This script was last ran on ' + now.strftime('%A %B %d, %Y') ))
```
