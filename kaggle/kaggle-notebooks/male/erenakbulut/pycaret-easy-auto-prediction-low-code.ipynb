# Read csv files

```python
import pandas as pd

train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
test_df = pd.read_csv('/kaggle/input/titanic/test.csv')
```

# Install llvmlite, is needed for this usecase

```python
!pip install llvmlite --ignore-installed
```

# Install PyCaret

```python
!pip install pycaret
```

# Import PyCaret and setup experiment

```python
from pycaret.classification import *

clf = setup(data = train_df, target = 'Survived')
```

# Compare models

```python
compare_models()
```

# Create a model using the best from above.

```python
rg_model = create_model("ridge")
```

# Tune our model

```python
tuned_rgmodel = tune_model(rg_model)
```

# Get predictions

```python
predictions = predict_model(tuned_rgmodel, data=test_df)
```

# Extract predictions

```python
submission_df = predictions[['Label']]
```

```python
submission_df.rename(columns={'Label':'Survived'}, inplace=True)
```

```python
submission_df['PassengerId'] = test_df['PassengerId']
```

# Create submission df

```python
submission_df = submission_df[['PassengerId', 'Survived']]
```

# Write to the local

```python
submission_df.to_csv('submission.csv', index=False)
```
