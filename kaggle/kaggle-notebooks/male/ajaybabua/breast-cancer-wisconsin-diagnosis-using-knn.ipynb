# **We are dropping columns – ‘id’ and ‘Unnamed: 32’ as they have no role in prediction**

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

# **Encoding the diagnosis column**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['diagnosis'] = encoder.fit_transform(df['diagnosis'])
df['diagnosis'].unique()
```

```python
import seaborn as sns

sns.lmplot(x = 'radius_mean', y = 'texture_mean', hue = 'diagnosis', data = df)
```

```python
sns.lmplot(x ='smoothness_mean', y = 'compactness_mean',data = df, hue = 'diagnosis')
```

# **Input and Output data**

```python
x = df.drop(['diagnosis'], axis = 1)
y = df['diagnosis']
```

```python
x
```

```python
y
```

# Splitting data into Train and Test in the ratio of 70:30

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30,train_size = 0.70, random_state = 0)
```

# Fitting train data's into KNeighborsClassifier()

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 13)

knn.fit(x_train, y_train)
```

# Finding the Score to the Model

```python
score = knn.score(x_test,y_test)
score
```

# Predicting the X_test data to the Model

```python
y_pred = knn.predict(x_test)
```

# Finding the Root Mean Square Error(RMSE) to the Model

```python
from sklearn.metrics import mean_squared_error

print(f'RMSE = {np.sqrt(mean_squared_error(y_test, y_pred))}')
```
