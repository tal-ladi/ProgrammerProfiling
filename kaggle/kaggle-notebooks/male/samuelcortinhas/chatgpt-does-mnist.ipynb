```python
# import necessary packages
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D

# load the MNIST dataset
train_data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test_data = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')

# extract the labels and pixel values from the dataset
train_labels = train_data['label']
train_pixels = train_data.drop('label', axis=1)
test_pixels = test_data

# reshape the pixel values into 28x28 arrays
train_pixels = train_pixels.values.reshape((-1, 28, 28, 1))
test_pixels = test_pixels.values.reshape((-1, 28, 28, 1))

# normalize the pixel values between 0 and 1
train_pixels = train_pixels / 255.0
test_pixels = test_pixels / 255.0

# create a sequential model
model = Sequential()

# add a convolutional layer with 32 filters and a 3x3 kernel
model.add(Conv2D(32, (3,3), padding='same', activation='relu', input_shape=(28, 28, 1)))

# add a max pooling layer with a 2x2 pool size
model.add(MaxPooling2D(pool_size=(2, 2)))

# add a dropout layer with a rate of 0.25
model.add(Dropout(0.25))

# flatten the output of the previous layers
model.add(Flatten())

# add a dense layer with 128 neurons and a ReLU activation
model.add(Dense(128, activation='relu'))

# add another dropout layer with a rate of 0.5
model.add(Dropout(0.5))

# add a final dense layer with 10 neurons (one for each digit) and a softmax activation
model.add(Dense(10, activation='softmax'))

# compile the model with categorical crossentropy loss and Adam optimization
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model using the training data
model.fit(train_pixels, train_labels, epochs=10, batch_size=128)

# evaluate the model on the test data
#test_loss, test_acc = model.evaluate(test_pixels, test_labels)
#print('Test accuracy:', test_acc)
```

The model is quite basic but it **performs relatively well - 98%**. It is interesting to see it chose to add **only one convolutional layer** as it is very standard to put two or more convolutional layers together.

```python
# Make predictions
preds = model.predict(test_pixels)

# Save predictions to csv
submission = pd.read_csv('/kaggle/input/digit-recognizer/sample_submission.csv')
submission['Label'] = np.argmax(preds, axis=1)
submission.to_csv('submission.csv', index=False)
```
