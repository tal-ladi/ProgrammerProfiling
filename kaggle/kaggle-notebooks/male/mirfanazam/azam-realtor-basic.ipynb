```python
feature_names = ["MSSubClass", "MSZoning", "LotArea", "OverallQual", "OverallCond", "YearBuilt","1stFlrSF","2ndFlrSF"
                 ,"LowQualFinSF", "TotalBsmtSF", "GrLivArea","FullBath", "HalfBath","BedroomAbvGr","TotRmsAbvGrd"
                 ,"OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch","MoSold", "YrSold",'YearRemodAdd'
                 ,'KitchenAbvGr','Fireplaces','WoodDeckSF','PoolArea','MiscVal','LotFrontage','Street','LandContour'
                 ,'Alley','Neighborhood']
```

# **Feature Engineering**

```python
def fill_missing_ex(df):
    for column in df:
        if df[column].dtype == 'int64':
            df[column].fillna(0, inplace = True)
        elif df[column].dtype == 'float64':
            df[column].fillna(0.0, inplace = True)
        else:
            df[column].fillna('U', inplace = True)
    return df

fill_missing_ex(home_data_train)
fill_missing_ex(home_data_test)
```

```python
from sklearn import preprocessing

def encode_features(df):
    le = preprocessing.LabelEncoder()
    for column in df:
        if df[column].dtype == 'object':
            le.fit(df[column])
            df[column] = le.transform(df[column])
    return df

encode_features(home_data_train)
encode_features(home_data_test)
```

# **Build Model**

```python
X = home_data_train[feature_names]
X_test = home_data_test[feature_names]
```

```python
y = home_data_train.SalePrice
```

```python
iowa_model = GradientBoostingRegressor(random_state = 1)

iowa_model.fit(X,y)
```

# **Model Validation**

```python
predictions = iowa_model.predict(X)
y_train = np.array(predictions).tolist()
y_train = pd.DataFrame(y_train)
y_train.columns = ['SalePrice']
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y, y_train)
```

# **Predict Target Sale Price**

```python
predictions = iowa_model.predict(X_test)
```

```python
df = home_data_test[['Id']]

y_test = np.array(predictions).tolist()
y_test = pd.DataFrame(y_test)
y_test.columns = ['SalePrice']

final_prediction = pd.concat([df, pd.DataFrame(y_test)], axis=1)
```
