```python
dataset['HomePlanet']=dataset.HomePlanet.fillna(dataset.HomePlanet.mode()[0])
```

```python
dataset.isnull().sum()
```

```python


from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
dataset['CryoSleep']=l1.fit_transform(dataset['CryoSleep'])
dataset['Cabin']=l1.fit_transform(dataset['Cabin'])
dataset['VIP']=l1.fit_transform(dataset['VIP'])
```

```python
dataset.isnull().sum()
```

```python
dataset['Destination'].max
```

```python
# print(dataset['CryoSleep'].mode())
# print(dataset['CryoSleep'].mean())
# print(dataset['CryoSleep'].median())
dataset['CryoSleep'].fillna(value=dataset['CryoSleep'].mean(),inplace=True)
dataset['Destination'].fillna(value='TRAPPIST-1e',inplace=True)
dataset['Age'].fillna(value=dataset['Age'].mean(),inplace=True)
dataset['RoomService'].fillna(value=dataset['RoomService'].mean(),inplace=True)
dataset['FoodCourt'].fillna(value=dataset['FoodCourt'].mean(),inplace=True)
dataset['ShoppingMall'].fillna(value=dataset['ShoppingMall'].mean(),inplace=True)
dataset['Spa'].fillna(value=dataset['Spa'].mean(),inplace=True)
dataset['VRDeck'].fillna(value=dataset['VRDeck'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=['PassengerId','Cabin','Name'],axis=1,inplace=True)
```

```python
x = dataset.iloc[: ,:-2].values
y = dataset.iloc[: ,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
# x.drop(columns=[3],axis=1,inplace=True)
x = np.delete(x, 3, axis=1)
```

```python
x = pd.DataFrame(x)
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
x[0:3,:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
```

```python
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
