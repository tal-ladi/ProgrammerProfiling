```python
dataset = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv')
dataset.head()
```

```python
from pycaret.regression import *
```

```python
data = dataset.sample(frac=0.9, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
exp_reg101 = setup(data = data, target = 'SalePrice', session_id=123,categorical_imputation= 'mode',
                  normalize = True, transformation = True, transform_target = True, numeric_imputation= 'mean', remove_outliers= True,outliers_threshold=0.05, remove_multicollinearity= True,feature_selection = True,feature_selection_threshold = 0.4, feature_selection_method='classic')
```

```python
best = compare_models(exclude = ['ransac'])
```

```python
huber = create_model('huber')
```

```python
print(huber)
```

```python
tuned_huber = tune_model(huber)
```

```python
plot_model(tuned_huber)
```

```python
plot_model(tuned_huber, plot = 'error')
```

```python
plot_model(tuned_huber, plot='feature')
```

```python
predict_model(tuned_huber)
```

```python
final_huber = finalize_model(tuned_huber)
print(final_huber)
```

```python
predict_model(final_huber)
```

```python
unseen_predictions = predict_model(final_huber, data=data_unseen)
unseen_predictions.head()
```

```python
from pycaret.utils import check_metric
check_metric(unseen_predictions.SalePrice, unseen_predictions.Label, 'R2')
```

```python
save_model(final_huber,'Final Huber Model 21Jan2021')
```
